### Zadanie 5

Rozwiązać równanie w liczbach zespolonych:

a) $z^2+z+1=0;$

b) $z^2+(1+4i)z-(5+i)=0;$

c) $z^4-2z^2+4=0;$

d) $z^4+(15+7i)z^2+(8-15i)=0.$

Zaznaczyć rozwiązania na płaszczyźnie.

```python
#złe
eq = z**2 + z + 1 == 0
solution = solve([
    eq.lhs().real() - eq.rhs().real() == 0,
    eq.lhs().imag() - eq.rhs().imag() == 0
], [x, y])
p = [s for s in solution if s[0].rhs().is_real() and s[1].rhs().is_real()]
pts = [(sol[0].rhs(), sol[1].rhs()) for sol in p]

rys = points(pts, size=30, color='red')
rys
```
