Korzystając z: 1) definicji macierzy odwrotnej; 2) operacji elementarnych, znaleźć macierz $A^{-1}$, jeżeli:

a)
$$
\begin{equation*}
    A =
    \left[\begin{array}{rr}
    3 & 1\\
    5 & 2
    \end{array}\right]

$$

b)
$$
\begin{equation*}
    A =
    \left[\begin{array}{rr}
    4 & 2\\
    2 & 1
    \end{array}\right]
\end{equation*}
$$

c)
$$
\begin{equation*}
    A =
    \left[\begin{array}{rrr}
    1 & 0 & 2\\
    2 & 1 & 1\\
    2 & 1 & 0
    \end{array}\right]
\end{equation*}
$$

```python
p,q,r,t = var('p,q,r,t')
a = matrix([[3,1],[5,2]])
b = matrix([[p, q],[r,t]])
e = a.parent().one()
solution = matrix(nrows=2, ncols=2, entries=[s.rhs() for s in solve([c == 0 for c in (a*b - e).list()], [p,q,r,t])[0]])
solution
```

```python
(a.augment(e, subdivide=True)).echelon_form()
```

```python
a.inverse()
a^(-1)
```
