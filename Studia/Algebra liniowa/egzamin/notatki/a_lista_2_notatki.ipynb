### Zadanie 3
---

Rozwiązać podane układy równań metodą kolumn jednostkowych:

a)
$$
\begin{equation*}
\left\{
\begin{array}{rrr}
2x-y+z & = & 1 \\
-4x-12y+z & = & 2\\
3x+3y+z & = & 3
\end{array}
\right.
\end{equation*}
$$

```python
A = matrix(3,3,[2,-1,1,-4,-12,1,3,3,1])
b = vector([1,2,3])
U = A.augment(b,subdivide=true)
show(U)
U = U.rref()
U
```

b)
$$
\begin{equation*}
\left\{
\begin{array}{rrr}
4x_1+3x_2+5x_3+7x_4 & = & 2 \\
2x_1-x_2+x_3+3x_4 & = & 4\\
x_1+2x_2+2x_3+2x_4 & = & -1\\
3x_1+x_2+3x_3+5x_4 & = & 3
\end{array}
\right.
\end{equation*}
$$

```python
A = matrix(4,4,[4,3,5,7,2,-1,1,3,1,2,2,2,3,1,3,5])
b = vector([2,4,-1,3])
U = A.augment(b,subdivide=true)
show(U)
U = U.rref()
show(U)
```

c)
$$
\begin{equation*}
\left\{
\begin{array}{rrr}
3x+2y+z-t & = & 0 \\
5x-y+z+2t & = & -4\\
7x+8y+z-7t & = & 6\\
x-y+z+2t & = & 4
\end{array}
\right.
\end{equation*}
$$

```python
A = matrix(4,4,[3,2,1,-1,5,-1,1,2,7,8,1,-7,1,-1,1,2])
b = vector([0,-4,6,4])
U = A.augment(b,subdivide=true)
show(U)
U = U.rref()
show(U)
```
