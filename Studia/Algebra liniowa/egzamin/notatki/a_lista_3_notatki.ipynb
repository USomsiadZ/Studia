a) $\sigma\tau,\quad$
b) $\tau\sigma,\quad$
c) $\sigma^4,\quad$
d) $\tau^{-1},\quad$
e) $\tau^{2021}$,

jeżeli
$$
\begin{equation*}
\sigma=\begin{pmatrix}
1 & 2 & 3 & 4 & 5 \\
2 & 5 & 3 & 1 & 4
\end{pmatrix},\qquad
\tau=
\begin{pmatrix}
1 & 2 & 3 & 4 & 5 \\

$$
\begin{equation*}
\sigma^{-26}\tau\sigma =

9 & 1 & 4 & 2 & 6 & 5 & 3 & 8 & 7
\end{pmatrix}

$$
