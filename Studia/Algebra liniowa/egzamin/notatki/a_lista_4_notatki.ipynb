### Zadanie 3
---
Znaleźć liczby rzeczywiste $x,y$ spełniające równanie:

a) $(3-i)x^2-(3+2i)x-(1-i)y=13-10i;$

b) $(2+ix)(y-4i)=14-2i;$

c) $\displaystyle\frac{1+xi}{y-2i}=-1+2i.$

```python
x,y = var('x,y', domain='real')
z = x+y*i

## a
eq = (3-i)*x**2 - (3+2*i)*x - (1-i)*y == 13-10*i
solve([eq.lhs().real() == eq.rhs().real(), eq.lhs().imag() == eq.rhs().imag()], [x,y])
```
