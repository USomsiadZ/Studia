### Zadanie 1
Narysować na płaszczyźnie trójkąt foremny, którego jeden z wierzchołków leży w punkcie $(3,5)$. Zakładamy, że środek trójkąta znajduje się w punkcie $(2,2)$.

```python
z = var('z', domain='complex')
u = 2+2*I
polygon([s.rhs() + u for s in solve([z**3 == ((3+5*I) - u)**3], [z])], fill=False)
```

### Zadanie 2
Trójkąt foremny o jednym z wierzchołków w punkcie $(-2,5)$ i środku w początku układu współrzędnych obrócić o kąt $\frac{\pi}{10}$ zgodnie z ruchem wskazówek zegara.

```python
z = var('z', domain='complex')
w = e**(i*(-pi/10))
p1 = polygon([s.rhs()  for s in solve([z**3 == (-2+5*I)**3], [z])], fill=False)
p2 = polygon([s.rhs()*w  for s in solve([z**3 == (-2+5*I)**3], [z])], fill=False, color='red')
p1 + p2
```

### Zadanie 3
Kwadrat o jednym z wierzchołków w punkcie $(0,5)$ i środku w początku układu współrzędnych obrócić o kąt $\frac{\pi}{6}$ przeciwnie do ruchu wskazówek zegara. Skorzystać z macierzy obrotu.

```python
z = var('z', domain='complex')
alpha = pi/6
a = matrix([[cos(alpha), -sin(alpha)], [sin(alpha), cos(alpha)]])
p1 = polygon([s.rhs() for s in solve([z**4 == (0+5*I)**4], [z])], fill=False)
p2 = polygon([a*vector([s.rhs().real(), s.rhs().imag()]) for s in solve([z**4 == (0+5*I)**4], [z])], color='red', fill=False)
p1 + p2
```
