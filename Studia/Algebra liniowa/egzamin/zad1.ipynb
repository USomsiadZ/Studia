```python
x, y = var('x, y')

f1 = x^2 + y^2 < 16
f2 = (x - 3)^2 + (y - 2)^2 > 9
f3 = x > 1

zakres_x = (-4, 5)
zakres_y = (-6, 6)

A = region_plot(
    [f1, f2, f3],
    zakres_x,
    zakres_y,
    incol='lightblue',
    bordercol='blue',
    plot_points=500,
    aspect_ratio=1
)

granica1 = implicit_plot(x^2 + y^2 == 16, zakres_x, zakres_y, color='red', linestyle='dashed')
granica2 = implicit_plot((x - 3)^2 + (y - 2)^2 == 9, zakres_x, zakres_y, color='green', linestyle='dashed')
granica3 = plot(1, (x, -4, 5), color='black', linestyle='dashed')

show(A + granica1 + granica2 + granica3, title='Wykres zbioru A: |z|<4 & |z-3-2i|>3 & Re z > 1')
```
