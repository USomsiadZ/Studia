```python
z, w = var('z w')

z = 3 - 2*I
w = -2 + 4*I

im_z = z.imag()
abs_w = abs(w)
re_w = w.real()

wyrazenie = (im_z + I*abs_w) / (re_w - z)
wynik = wyrazenie.full_simplify()
wynik_re = wynik.real().full_simplify()
wynik_im = wynik.imag().full_simplify()

print("\nWynik")
print(f"{wynik_re} + {wynik_im}*I")
```
