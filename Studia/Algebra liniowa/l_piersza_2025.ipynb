### Zadanie 2
---

Rozwiązać podane układy równań metodą eliminacji Gaussa:

a)
$$
\begin{equation*}
\left\{
\begin{array}{rrr}
3x+2y+4z+t & = & 1 \\
3x+y+6z+4t & = & 2\\
2x+2y+2z+2t & = & 0
\end{array}
\right.
\end{equation*}
$$

b)
$$
\begin{equation*}
\left\{
\begin{array}{rrr}
x_1+2x_2+3x_3+5x_4 & = & 3 \\
2x_1+4x_2-x_3-x_4 & = & 2\\
x_1+2x_2+x_3+x_4 & = & 1\\
2x_1+4x_2-3x_3-5x_4 & = & 0
\end{array}
\right.
\end{equation*}
$$

c)
$$
\begin{equation*}
\left\{
\begin{array}{rrr}
2x_1+4x_2-2x_3+4x_4 & = & 6 \\
3x_1+2x_2+x_3-2x_4 & = & 1\\
3x_1+4x_2-x_3+2x_4 & = & 5
\end{array}
\right.
\end{equation*}
$$
