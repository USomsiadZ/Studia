```python
p, q, r, t = var('p,q,r,t')
a = matrix(nrows=2, ncols=2, entries=[3,1,5,2])
b = matrix(nrows=2, ncols=2, entries=[p,q,r,t])

o = a.parent().one()

x = matrix(nrows=2, ncols=2, entries=[s.rhs() for s in solve([c == 0 for c in (a*b - o).list()], [p,q,r,t])[0]])

(a*x == x*a) and x*a == identity_matrix(2)
x
```
