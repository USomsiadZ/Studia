**Dygresja**:

Jednostka urojona $i$ to <u>liczba zespolona</u> o następującej własności:

$$i^2 = -1.$$

Oznacza to, że kwadrat liczby zespolonej <u>nie musi być nieujemny</u> (liczb zespolonych w ogóle <u>nie możemy porównywać</u>). Co więcej, równanie:

$$x^2 + 1 = 0$$

ma <u>2 rozwiązania zespolone</u> (mimo że $\Delta < 0$). Są to liczby należące do zbioru $\{ -i, i \}$.

---

W tzw. <u>ciele liczb zespolonych</u> równanie wielomianowe $n$-tego stopnia zawsze ma <u>dokładnie $n$ rozwiązań zespolonych</u> (licząc z krotnościami).

---

Liczby zespolone można utożsamiać z <u>punktami na płaszczyźnie</u>. Przykładowo, liczbę zespoloną $3 + 2i$ identyfikujemy z punktem $(3, 2)$ w układzie współrzędnych.

---

Jednostka urojona występuje we wzorze, nazywanym często <u>najpiękniejszym wzorem matematyki</u>, czyli:

$$e^{i \pi} + 1 = 0,$$

łączącym 5 ważnych stałych matematycznych: $e$, $i$, $\pi$, 1 oraz 0.

**Dygresja**:

<u>Funkcje cyklometryczne</u> (arcusy) to funkcje odwrotne do odpowiednich <u>funkcji trygonometrycznych</u>. Przykładowo, $y = \arccos x$ jest funkcją odwrotną do fragmentu&nbsp;funkcji $y = \cos x$, gdzie dziedziną jest zbiór $[0, \pi]$ (mówimy, że funkcja jest <u>zawężona</u> do przedziału $[0, \pi]$ i piszemy $y = \cos x |_{[0, \pi]}$).

---

Wykres <u>funkcji odwrotnej</u> powstaje poprzez <u>symetrię względem prostej</u> $y=x$. Na przykład, na rysunku poniżej przedstawiony jest proces powstawania funkcji $y = \operatorname{arctg} x$. Kolorem <u>czarnym</u> zaznaczono fragment funkcji $y = \operatorname{tg} x |_{\left( -\frac{\pi}{2}, \frac{\pi}{2} \right)}$, a kolorem <u>niebieskim</u> – przekątną $y=x$. Funkcja $y = \operatorname{arctg} x$, narysowana kolorem <u>czerwonym</u>, jest lustrzanym odbiciem wykresu czarnego względem niebieskiej prostej:

```python
tg = plot( tan(x), (x, -pi/2, pi/2), color='black', ymin=-10, ymax=10, legend_label=r'$y=\operatorname{tg} x$' )
prosta = plot( x, (x, -10, 10), color='blue', linestyle='--', legend_label='$y=x$' )
arctg = plot( arctan(x), (x, -10, 10), color='red', legend_label=r'$y=\operatorname{arctg} x$' )

tg + prosta + arctg
```

### 1.1.2. Wyrażenia algebraiczne<a class="anchor" id="1-2"></a>

System **SageMath** pozwala tworzyć <u>wyrażenia symboliczne</u>, operować na zmiennych <u>bez przypisywania im konkretnych wartości liczbowych</u> oraz przeprowadzać dokładne obliczenia.

Zmienna `x` <u>traktowana jest jako symbol domyślnie</u>, co pozwala od razu wykonywać działania na wyrażeniach algebraicznych:

```python
(2*x)^2 + x + 3*x + 1
```

Jednak gdy chcemy użyć innej nazwy zmiennej, należy ją najpierw <u>zadeklarować jako symbol</u>. W przeciwnym wypadku pojawi się następujący błąd:

```python
z + 3*z
```

Najważniejsza informacja znajduje się <u>na samym dole</u> – wskazuje ona, co należy poprawić. W naszym przypadku próbowaliśmy wykonać działanie na wyrażeniach algebraicznych, ale nie zdefiniowaliśmy zmiennej `z` jako <u>symbolu</u>. Należy to zrobić za pomocą funkcji **var()**:

```python
var('z')

z + 3*z
```

Można również deklarować jako symbole <u>wiele zmiennych jednocześnie</u>:

```python
var('s t u')

wyrazenie = -s + 3*t + 2*u
```

i następnie wykonywać na nich działania:

```python
wyrazenie + 2*s
```

---

**Uwaga** (najczęściej popełniane błędy):

1. **Brak operatora gwiazdki przy mnożeniu**

Poprawny jest zapis `3*x`, `5*x^2` itd. Natomiast instrukcja `2x` spowoduje błąd `invalid syntax` lub `invalid decimal literal`, więc jeśli widzimy taki komunikat, należy <u>szukać w kodzie brakujących gwiazdek</u>:

```python
2x
```

2. **Brak deklaracji zmiennych jako symbole**

Stosowanie niezdefiniowanych zmiennych generuje błąd `Name 'var' Is Not Defined`, z którym już wcześniej się spotkaliśmy. Co więcej, jeżeli przypiszemy do zmiennej jakąś <u>wartość liczbową</u>, na przykład `x=7`, to od tej pory będzie ona traktowana jako <u>liczba</u>, a nie <u>symbol</u>. Aby ponownie wykonywać działania na wyrażeniach algebraicznych, należy zastosować polecenie `var('x')`:

```python
var('x')
print( 'Typ dla symbolu:', type(x) )

x = 7
print( 'Typ po przypisaniu liczby:', type(x) )

# jeżeli chcemy uzyskać 3*x, trzeba ponownie zadeklarować zmienną x jako symbol za pomocą var('x')
x + 2*x
```

3. **Stosowanie nazw wykorzystywanych w Sage**

<u>Nadpisanie wbudowanych nazw</u>, na przykład jako `e = 2` lub `cos = 1.5`, sprawia, że stała Eulera i funkcja cosinus będą odtąd traktowane jak <u>zwykłe liczby</u>. W rezultacie późniejsze użycie tych nazw spowoduje błąd `object is not callable` lub błędny wynik:

```python
cos = 1.5
cos(0)
```

Można to naprawić poprzez <u>restart kernela</u> (**Kernel** → **Restart Kernel**) albo za pomocą polecenia **reset()**:

```python
reset()

cos(0)
```

4. **Brak nawiasów, gdy istotna jest kolejność działań**

Wyrażenia w liczniku i mianowniku ułamka zawsze są <u>domyślnie w nawiasie</u>. Dlatego podczas definiowania ułamka:

$$\frac{x+2}{x-1}$$

należy koniecznie <u>dodać w kodzie nawiasy</u>:

```python
wyrazenie = (x + 2) / (x - 1)
wyrazenie
```

Jeżeli napiszemy `x + 2 / x - 1`, to <u>obowiązywać będzie kolejność działań</u>, więc w wyniku otrzymamy błędny rezultat:

$$x+\frac{2}{x}-1.$$

Poniżej przedstawiony jest niepoprawny zapis:

```python
wyrazenie = x + 2 / x - 1  # źle!
wyrazenie
```

---

Zauważmy, że wyrażenia algebraiczne również domyślnie zapisywane są w <u>najprostszej postaci</u> (najczęściej jest to postać iloczynowa). Jednak możemy wymusić określone zachowanie wyrażeń za pomocą odpowiednich funkcji:

1. **expand()** – zapisuje wyrażenie w <u>postaci ogólnej</u>:

```python
wyrazenie = (x+2)^3
print(wyrazenie)  # warto wypisywać wyrażenia, aby się upewnić czy dostaliśmy to, o co nam chodziło

wyrazenie.expand()
```

2. **factor()** – zapisuje wyrażenie w <u>postaci iloczynowej</u>:

```python
wyrazenie = x^2 - x - 6
print(wyrazenie)

wyrazenie.factor()
```

3. **simplify()** – <u>upraszcza</u> wyrażenie (można stosować funkcje **simplify_rational()** dla <u>wyrażeń wymiernych</u> lub **simplify_trig()** dla <u>funkcji trygonometrycznych</u>):

```python
wyrazenie = (x^3 - 3*x - 2) / (x^2 - x - 2)
print(wyrazenie)

# skracanie ułamka
wyrazenie.simplify_rational()
```

W zależności od rodzaju wyrażenia możemy używać innych funkcji upraszczających lub wypróbować **simplify_full()** w celu ogólnego uproszczenia:

```python
wyrazenie = sin(x)^2 + cos(x)^2
print(wyrazenie)

# trygonometria
wyrazenie.simplify_trig()
```

Zobaczymy teraz, jak można wyznaczać <u>wartość wyrażenia</u> dla określonej zmiennej.

**Uwaga**:

W **Sage** występuje rozróżnienie między <u>wyrażeniem</u> a <u>funkcją</u>. <u>Funkcje</u> definiujemy poprzez dodanie nawiasów, w których określamy zmienne zależne. <u>Wszystkie te zmienne zostaną od razu potraktowane jako symbole</u>, więc nie ma konieczności korzystania z funkcji **var()**:

```python
f(n) = n^2 + 2*n - 2
```

W przeciwieństwie do wyrażenia, przy obliczaniu wartości funkcji dla danego argumentu, na przykład $n=3$, <u>nie musimy stosować zapisu ze znakiem równości</u> – wystarczy przekazać poszczególne argumenty w odpowiedniej kolejności:

```python
f(3)
```

Funkcja, w odróżnieniu od wyrażenia, jest wyświetlana <u>razem ze strzałką</u>:

```python
f
```

Zamiast tzw. <u>surowego tekstu</u> uzyskanego poprzez wykorzystanie litery `r`, można też stosować podwójny backslash (`\\`).

## 1.2. Rysunki<a class="anchor" id="2"></a>

### 1.2.1. Rodzaje funkcji graficznych<a class="anchor" id="2-1"></a>

Przeanalizujemy podstawowe funkcje dostępne w **Sage** służące do generowania <u>wykresów funkcji</u> oraz rysowania <u>figur geometrycznych</u>:

1. <u>Wykres funkcji</u> – **plot()**

Aby narysować wykres, jako pierwszy argument funkcji **plot()** przekazujemy <u>funkcję</u>, <u>wyrażenie</u> lub <u>wzór funkcji</u>, której wykres chcemy otrzymać. Domyślnie <u>dziedziną</u> jest przedział $(-1, 1)$:

```python
plot(x^3)
```

Aby <u>zmienić dziedzinę</u>, piszemy:

```python
plot( x^3, (x, -2, 2) )
```

<u>Zbiór wyświetlanych wartości</u> można kontrolować za pomocą argumentów **ymin** i **ymax**. Przykładowo, narysujemy wykres funkcji $f(x)=\sin x$ dla <u>argumentów</u> z przedziału $(-\pi, \pi)$ z <u>wartościami</u> ograniczonymi do przedziału $(-2, 2)$:

```python
plot( sin(x), (x, -pi, pi), ymin=-2, ymax=2 )
```

2. <u>Obszar</u> – **region_plot()**

Funkcja **region_plot()** służy do zaznaczania <u>obszaru na płaszczyźnie</u> określonego za pomocą <u>ograniczeń</u> podanych w nawiasach kwadratowych. Warunki oddzielone przecinkami oznaczają ich <u>część wspólną</u>:

```python
var('x y')

region_plot( [x > 1, y > 2, y < 5 - x], (x, 0, 4), (y, 1, 5) )
```

Początkowo uzyskany obszar nie prezentuje się jeszcze zbyt ładnie, dlatego warto poprawić jego wygląd, modyfikując <u>kolory brzegu i wypełnienia</u> oraz <u>grubość linii</u> poprzez zmianę odpowiednich argumentów:

```python
region_plot( [x > 1, y > 2, y < 5 - x], (x, 0, 4), (y, 1, 5), incol='cyan', bordercol='black', borderwidth=2 )
```

3. <u>Punkt</u> – **point()**

Aby narysować <u>punkt na płaszczyźnie</u>, należy przekazać jego współrzędne do funkcji **point()**. Podobnie jak inne funkcje graficzne, **point()** posiada <u>własny zestaw argumentów</u>. Poniżej znajduje się kod, który jest <u>skuteczny w przypadku większości rysunków</u>:

```python
point( (2, 3), color='red', size=30, zorder=5 )  # kolor, rozmiar i kolejność wyświetlania
```

4. <u>Linia</u> – **line()**

Aby narysować <u>linię łączącą dwa punkty</u>, umieszczamy w liście <u>współrzędne jej końców</u>:

```python
line( [(1, -1), (1, 4)], color='purple', thickness=3 )
```

5. <u>Wielokąt</u> – **polygon()**

Pierwszym argumentem funkcji **polygon()** jest lista zawierająca <u>współrzędne kolejnych wierzchołków</u> wielokąta. Ważne jest przekazanie ich <u>zgodnie z określoną kolejnością</u> – wzdłuż boków wielokąta:

```python
polygon( [(1, 1), (-1, 2), (2, 3), (4, 1)], color='cyan', edgecolor='black' )
```

Możemy również narysować wielokąt <u>pusty w środku</u>:

```python
polygon( [(1, 1), (2, 1), (1, 2)], fill=False )
```

Jeżeli przekażemy do którejkolwiek z powyższych funkcji <u>nieistniejący argument</u>, wyświetli się pomoc z podpowiedziami dotyczącymi ich nazw:

## 1.3. Równania<a class="anchor" id="3"></a>

### 1.3.1. Rozwiązywanie równań i nierówności<a class="anchor" id="3-1"></a>

Podstawowym narzędziem do rozwiązywania <u>równań i nierówności</u> w **Sage** jest funkcja **solve()**. Pierwszym argumentem tej funkcji jest <u>równość</u>, którą zapisujemy za pomocą operatora `==`. Drugi argument określa zmienną, względem której szukamy rozwiązania.

---

**Przykład**

Rozwiążemy równanie:

$$2x^2 - 5x = 3.$$

---

Stosujemy składnię **solve**(*równanie*, *zmienna*). W wyniku otrzymujemy <u>listę równości</u> będących rozwiązaniami równania:

```python
solve(2*x^2 - 5*x == 3, x)
```

**Przykład**

Rozwiążemy równanie:

$$x^3 + 3x^2 - 2x - 6 = 0.$$

---

Jeżeli po <u>prawej stronie równości znajduje się liczba 0</u>, możemy <u>pominąć znak równości</u>. Funkcja **solve()** domyślnie zakłada, że prawa strona jest równa 0:

```python
solve(x^3 + 3*x^2 - 2*x - 6, x)
```

W analogiczny sposób rozwiązujemy <u>nierówności</u>.

---

**Przykład**

Rozwiążemy nierówność:

$$x^2 - 2x - 3 > 0.$$

---

Tym razem również stosujemy funkcję **solve()**:

```python
solve(x^2 - 2*x - 3 > 0, x)
```

W **Sage** nawiasy kwadratowe oddzielone przecinkiem, na przykład `[[x < -1], [x > 3]]`, oznaczają <u>sumę przedziałów</u>, czyli zbiór $(-\infty, -1) \cup (3, +\infty)$. Natomiast zapis `[[x > -1, x < 3]]` rozumiemy jako <u>część wspólną przedziałów</u>, czyli zbiór $(-1, 3)$.

Zwracany wynik jest <u>listą</u>, więc możemy odwoływać się do poszczególnych rozwiązań za pomocą indeksów.

---

**Przykład**

Rozwiążemy poniższe równanie oraz podamy przybliżoną wartość pierwszego rozwiązania:

$$x^3 + 2x^2 = 3x + 6.$$

---

Najpierw wypisujemy <u>wszystkie rozwiązania</u> równania:

```python
rozw = solve(x^3 + 2*x^2 == 3*x + 6, x)
rozw
```

Następnie wybieramy <u>pierwsze rozwiązanie</u> z uzyskanej listy. Jest to <u>równość</u>, ponieważ występuje w niej charakterystyczny podwójny znak równości (`==`):

```python
rozw[0]
```

Nie możemy jeszcze obliczyć wartości przybliżonej pierwszego rozwiązania jako `rozw[0].n()`, ponieważ otrzymalibyśmy błąd. Najpierw należy wyodrębnić <u>prawą stronę</u> tej równości (**Right-Hand Side**) za pomocą funkcji **rhs()**:

```python
rozw[0].rhs()
```

Dopiero teraz podajemy wartość przybliżoną, korzystając z funkcji **n()**:

```python
rozw[0].rhs().n()
```

Możemy wypisać wszystkie rozwiązania, iterując w pętli **for** po liście zwracanej przez funkcję **solve()**.

---

**Przykład**

Rozwiążemy równanie:

$$x^4 - 1 = 0.$$

---

Stosujemy poniższy kod:

```python
for r in solve(x^4-1, x):
    show( r.rhs() )
```

Aby wprowadzić <u>założenia</u> dotyczące dziedziny równania, korzystamy z funkcji **assume()**. Poniżej zakładamy, że dziedziną jest zbiór <u>dodatnich liczb rzeczywistych</u>:

```python
assume(x, 'real')
assume(x > 0)

for r in solve(x^4-1, x):
    show( r.rhs() )
```

Funkcja **forget()** służy do <u>zapominania o założeniach</u>. Możemy również kontrolować zwracane wyniki w pętli za pomocą instrukcji warunkowej **if**. W **Sage** mamy następujące nazewnictwo zbiorów liczbowych:

* `NN` – zbiór liczb <u>naturalnych</u>;
* `ZZ` – zbiór liczb <u>całkowitych</u>;
* `QQ` – zbiór liczb <u>wymiernych</u>;
* `RR` – zbiór liczb <u>rzeczywistych</u>.

---

Warto zwrócić uwagę, że w poniższej pętli sprawdzamy jedynie czy <u>prawa strona równości</u> jest liczbą rzeczywistą. Zapis `r in RR` nigdy nie będzie prawdą, ponieważ <u>równość</u> `r` nie jest <u>liczbą</u>:

```python
forget()

for r in solve(x^4-1, x):
    if r.rhs() in RR:
        show(r)
```

### 1.3.2. Numeryczne rozwiązywanie równań<a class="anchor" id="3-2"></a>

Gdy równanie jest zbyt <u>skomplikowane</u>, możemy znaleźć jego rozwiązanie <u>numerycznie</u> w określonym przedziale, korzystając z funkcji **find_root()**.

---

**Przykład**

Rozwiążemy numerycznie równanie:

$$e^x + \ln x = x.$$

---

Funkcja **solve()** <u>nie zwraca wyniku</u>, ponieważ wprowadzone równanie jest zbyt trudne. Otrzymujemy jedynie nieco przekształconą wersję pierwotnej równości:

```python
rownanie = exp(x) + ln(x) == x

solve(rownanie, x)
```

W takiej sytuacji należy rozwiązać równanie <u>numerycznie</u>, korzystając ze składni **find_root**(*równanie*, *lewy*, *prawy*). Wówczas funkcja automatycznie zastosuje odpowiednią metodę obliczeniową i zwróci <u>jedno rozwiązanie</u> (lub informację o jego braku) w przedziale [*lewy*, *prawy*]. Jeżeli w&nbsp;zadanym przedziale jest <u>więcej pierwiastków</u>, trzeba odpowiednio <u>zawęzić przeszukiwany przedział</u>, aby znaleźć wszystkie rozwiązania:

```python
# do usuwania ostrzeżeń o przedawnieniu funkcji
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)

find_root(rownanie, 0, 1)
```

Skąd wiemy, że rozwiązania należy szukać akurat w przedziale $[0, 1]$? Warto najpierw <u>narysować wykresy funkcji</u> występujących po obu stronach równania, aby zorientować się, gdzie znajduje się potencjalne rozwiązanie:

```python
plot( [rownanie.lhs(), rownanie.rhs()], (x, 0, 2) )
```

Widzimy, że wykresy przecinają się w punkcie o pierwszej współrzędnej $x \approx 0.35$, więc spodziewamy się rozwiązania w przedziale $[0, 1]$. Na tej podstawie wybieramy wartości przekazane do funkcji **find_root()**.

---

**Przykład**

Rozwiążemy numerycznie równanie:

$$x^2 - 2x = \sin x.$$

---

Zaczynamy od <u>rysunku pomocniczego</u>, aby zlokalizować potencjalne rozwiązania:

```python
rownanie = x^2 - 2*x == sin(x)

plot( [rownanie.lhs(), rownanie.rhs()], (x, -1, 3) )
```

Na rysunku dostrzegamy <u>dwa punkty przecięcia</u>. Znajdują się one w przedziałach, na przykład, $[-1, 1]$ oraz $[2, 3]$. Wobec tego rozwiązania tego równania znajdujemy następująco:

```python
print( 'Pierwsze rozwiązanie:', find_root(rownanie, -1, 1) )
print( 'Drugie rozwiązanie:', find_root(rownanie, 2, 3) )
```

Pierwsza wartość jest liczbą równą około $2 \cdot 10^{-16}$, więc jest to w przybliżeniu 0.

---

Ponownie, więcej dostępnych informacji dotyczących <u>rozwiązywania równań</u> można znaleźć w [dokumentacji równań i nierówności](https://doc.sagemath.org/html/en/reference/calculus/sage/symbolic/relation.html) w **Sage**.

## 1.4. Macierze<a class="anchor" id="4"></a>

### 1.4.1. Definiowanie macierzy<a class="anchor" id="4-1"></a>

<u>Macierz</u> to tablica liczb. Aby odczytać <u>wymiary macierzy</u>, podajemy <u>najpierw liczbę wierszy</u>, a później liczbę kolumn.

---

**Przykład**

Zdefiniujemy macierz o wymiarach $3\times4$:

$$\textbf{M} = \begin{bmatrix} 2 & -1 & 5 & 3 \\ -3 & 0 & 1 & 1 \\ -1 & 4 & 0 & -1 \end{bmatrix}.$$

---

Aby utworzyć macierz w **Sage**, korzystamy z funkcji **matrix()**. Kolejne <u>wiersze</u> wpisujemy w oddzielnych nawiasach kwadratowych:

```python
M = matrix( [[2, -1, 5, 3], [-3, 0, 1, 1], [-1, 4, 0, -1]] )
show(M)
```

Można również przekazać kolejne <u>elementy w liście</u>, zaznaczając ile macierz powinna mieć wierszy (`nrows`) lub kolumn (`ncols`):

```python
M = matrix( [2, -1, 5, 3, -3, 0, 1, 1, -1, 4, 0, -1], nrows=3 )
M
```

Liczbę wierszy i kolumn odczytujemy odpowiednio za pomocą funkcji **nrows()** i **ncols()**:

```python
print( 'Liczba wierszy:', M.nrows() )
print( 'Liczba kolumn:', M.ncols() )
```

<u>Segmentacja macierzy</u> przebiega podobnie, jak w przypadku listy w języku **Python**.

---

**Przykład**

Zaprezentujemy metody segmentacji macierzy:

$$\textbf{C} = \begin{bmatrix} c_{11} & c_{12} & c_{13} \\ c_{21} & c_{22} & c_{23} \\ c_{31} & c_{32} & c_{33} \end{bmatrix} = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}.$$

---

Macierz $\textbf{C}$ można zdefiniować sprytnie, wykorzystując funkcję **range()**:

```python
C = matrix( range(1, 10), nrows=3 )
C
```

Do poszczególnych elementów $c_{ij}$ możemy odwoływać się za pomocą składni `C[i, j]` albo `C[i][j]`. Poniżej wypiszemy wartości elementów $c_{11}$ i $c_{23}$:

```python
print( 'Element c11:', C[0, 0] )
print( 'Element c23:', C[1][2] )
```

Drugi sposób zapisu, czyli `C[i][j]`, jest <u>bardziej restrykcyjny</u> – nie pozwala na nadpisywanie elementów macierzy. Zatem funkcjonuje podobnie jak <u>krotka</u> w języku **Python**:

```python
C[2][2] = 12
```

Natomiast pierwsza konwencja zapisu, czyli `C[i, j]`, pozwala już na poprawne wykonanie tej <u>operacji nadpisania</u>:

```python
C[2, 2] = 12
C
```

```python
print('Macierz 2x2 w lewym górnym rogu:')
show( C[:2, :2] )
print('Macierz 2x2 w prawym dolnym rogu:')
show( C[1:, 1:] )
```

### 1.4.2. Działania na macierzach<a class="anchor" id="4-2"></a>

Zbiór <u>wszystkich macierzy</u> o wymiarach $m\times n$ i elementach ze zbioru $\mathbb{R}$ oznaczamy symbolem $\textbf{M}_{m\times n}(\mathbb{R})$.

---

**Przykład**

Weźmy dwie macierze $2\times 2$, czyli elementy zbioru $\textbf{M}_{2\times 2}(\mathbb{R})$:

$$\textbf{A} = \begin{bmatrix} 1 & 3 \\ 0 & 2 \end{bmatrix} \quad \text{oraz} \quad \textbf{B} = \begin{bmatrix} 2 & 3 \\ 4 & 1 \end{bmatrix}.$$

---

Możemy wykonywać następujące <u>operacje na macierzach</u>:

1. **Dodawanie**

Dwie macierze możemy <u>dodawać i odejmować</u> tylko wtedy, gdy mają <u>takie same wymiary</u>. Wówczas wykonujemy te operacje dla <u>odpowiednich elementów</u>:

$$\textbf{A} + \textbf{B} = \begin{bmatrix} 1 & 3 \\ 0 & 2 \end{bmatrix} + \begin{bmatrix} 2 & 3 \\ 4 & 1 \end{bmatrix} = \begin{bmatrix} 1 + 2 & 3 + 3 \\ 0 + 4 & 2 + 1 \end{bmatrix} = \begin{bmatrix} 3 & 6 \\ 4 & 3 \end{bmatrix}.$$

```python
A = matrix( [[1, 3], [0, 2]] )
B = matrix( [[2, 3], [4, 1]] )

A + B
```

2. **Odejmowanie**

Dodawanie macierzy jest działaniem <u>przemiennym</u> (czyli $\textbf{A}+\textbf{B}=\textbf{B}+\textbf{A}$). Natomiast przy odejmowaniu istotna jest kolejność:

$$\textbf{A} - \textbf{B} = \begin{bmatrix} 1 & 3 \\ 0 & 2 \end{bmatrix} - \begin{bmatrix} 2 & 3 \\ 4 & 1 \end{bmatrix} = \begin{bmatrix} 1 - 2 & 3 - 3 \\ 0 - 4 & 2 - 1 \end{bmatrix} = \begin{bmatrix} -1 & 0 \\ -4 & 1 \end{bmatrix}.$$

```python
A - B
```

3. **Mnożenie przez liczbę**

Mówimy też – mnożenie przez <u>skalar</u>. Skalar to <u>element ciała</u>, a przykłady ciał to zbiór <u>liczb rzeczywistych</u> $\mathbb{R}$ oraz zbiór  <u>liczb zespolonych</u> $\mathbb{C}$:

$$2\textbf{A} = 2 \cdot \begin{bmatrix} 1 & 3 \\ 0 & 2 \end{bmatrix} = \begin{bmatrix} 2 \cdot 1 & 2 \cdot 3 \\ 2 \cdot 0 & 2 \cdot 2 \end{bmatrix} = \begin{bmatrix} 2 & 6 \\ 0 & 4 \end{bmatrix}.$$

---

Mnożenie przez skalar jest <u>przemienne</u>, czyli dla dowolnej liczby $k$ oraz macierzy $\textbf{A}$ mamy:

$$k \cdot \textbf{A} = \textbf{A} \cdot k.$$

```python
2 * A
```

4. **Mnożenie macierzy**

Dwie macierze można pomnożyć tylko wtedy, gdy ich <u>środkowe wymiary są takie same</u>.

Oznacza to, że działanie mnożenia jest możliwe dla macierzy o wymiarach $m \times k$ oraz $k \times n$. W wyniku otrzymujemy wówczas macierz o wymiarach $m \times n$.

Poniżej przedstawiony jest schemat ułatwiający zapamiętanie sposobu mnożenia macierzy – <u>łączymy je wierzchołkami</u> i znajdujemy sumę iloczynów odpowiadających sobie elementów:

---

<center><img src="https://www.dropbox.com/scl/fi/p1fgq2jrrdy3mp324bgt5/macierz.png?rlkey=53i70m89a66nlt5dktri2a0b2&st=0c5da4ci&raw=1" alt="Mnożenie macierzy" width="300";></center>

---

<u>Mnożenie macierzy nie jest przemienne</u>. Zapiszemy to schematycznie jako:

$$\textbf{AB} \neq \textbf{BA}.$$

Co więcej, może się okazać, że jednego z tych działań nie można wykonać – na przykład, gdyby macierze $\textbf{A}$ i $\textbf{B}$ miały wymiary odpowiednio $2\times3$ i $3\times4$.

---

W **Sage** korzystamy z operatora `*`:

```python
A * B
```

5. **Potęgowanie**

Potęgowanie macierzy jest skróconym zapisem jej <u>mnożenia przez siebie samą</u>, czyli:

$$\textbf{A}^n = \underbrace{\textbf{A} \cdot \ldots \cdot \textbf{A}}_{n}.$$

---

Mamy:

$$\textbf{B}^2 = \textbf{B} \cdot \textbf{B} = \begin{bmatrix} 2 & 3 \\ 4 & 1 \end{bmatrix} \cdot \begin{bmatrix} 2 & 3 \\ 4 & 1 \end{bmatrix} = \begin{bmatrix} 2 \cdot 2 + 3 \cdot 4 & 2 \cdot 3 + 3 \cdot 1 \\ 4 \cdot 2 + 1 \cdot 4 & 4 \cdot 3 + 1 \cdot 1 \end{bmatrix} = \begin{bmatrix} 16 & 9 \\ 12 & 13 \end{bmatrix}.$$

---

Potęgować możemy tylko <u>macierz kwadratową</u>, czyli macierz o takiej samej liczbie wierszy i kolumn. W **Sage** niezmiennie stosujemy operator potęgowania (`^`):

```python
B^2
```

6. **Transpozycja**

Transpozycję macierzy wykonujemy poprzez zapisanie jej kolejnych <u>wierszy jako kolumn</u>. Macierz transponowana do macierzy o wymiarach $m \times n$ ma wymiar $n \times m$:

$$\textbf{A}^{\text{T}} = \begin{bmatrix} 1 & 3 \\ 0 & 2 \end{bmatrix}^{\text{T}} = \begin{bmatrix} 1 & 0 \\ 3 & 2 \end{bmatrix}.$$

---

W **Sage** wykonujemy tę operację za pomocą funkcji **transpose()** lub krócej jako:

```python
A.T
```

7. **Łączenie macierzy**

W języku **Sage** macierze można połączyć:

* <u>wzdłuż wierszy</u>, korzystając z funkcji **augment()**;
* <u>wzdłuż kolumn</u> za pomocą funkcji **stack()**.

---

Najpierw łaczymy macierze $\textbf{A}$ i $\textbf{B}$ <u>w poziomie</u>:

```python
A.augment(B)
```

a następnie <u>w pionie</u>:

```python
A.stack(B)
```

Na koniec przyjrzymy się niektórym macierzom charakterystycznym:

1. **Macierz zerowa**

<u>Macierzą zerową</u> nazywamy macierz złożoną <u>wyłącznie z zer</u>. Oznaczamy ją symbolem **0** i ma ona tę własność, że dla każdej macierzy $\textbf{A}$ zachodzi:

$$\textbf{A}+\textbf{0}=\textbf{0}+\textbf{A}=\textbf{A}.$$

---

W **Sage** można ją utworzyć za pomocą funkcji **zero_matrix()** – jako argumenty przekazujemy kolejno liczbę wierszy i kolumn docelowej macierzy:

```python
zero_matrix(3, 2)
```

2. **Macierz jednostkowa**

<u>Macierz jednostkowa</u>, oznaczana symbolem $\textbf{I}$, ma na tzw. <u>przekątnej głównej</u> jedynki, a w pozostałych miejscach zera. Dla dowolnej macierzy <u>kwadratowej</u> zachodzi:

$$\textbf{AI}=\textbf{IA}=\textbf{A}.$$

---

Korzystamy z funkcji **identity_matrix()**:

```python
identity_matrix(3)
```

3. **Macierz diagonalna**

<u>Macierz diagonalna</u> to <u>macierz kwadratowa</u>, w której wszystkie elementy spoza <u>przekątnej głównej</u> są równe zeru. Tworzymy ją za pomocą funkcji **diagonal_matrix()**, gdzie przekazujemy listę kolejnych elementów do umieszczenia na głównej przekątnej:

```python
diagonal_matrix( [1, 2, 4, 3] )
```

4. **Macierz blokowa**

<u>Macierz blokowa</u> to macierz, którą traktujemy jako podzieloną na prostokątne fragmenty, zwane <u>blokami</u>. Na przykład, macierz $4\times4$ można zapisać jako macierz blokową postaci:

$$\begin{bmatrix} \textbf{A} & \textbf{B} \\ \textbf{C} & \textbf{D} \end{bmatrix},$$

gdzie $\textbf{A}$, $\textbf{B}$, $\textbf{C}$ i $\textbf{D}$ są macierzami $2\times2$. Macierz tę tworzymy, sklejając mniejsze macierze za pomocą funkcji **block_matrix()**:

```python
A = matrix( [[1, 2], [3, 4]] )
B = matrix( [[5, 6], [7, 8]] )

block_matrix( [[A, B], [B, A]] )
```
