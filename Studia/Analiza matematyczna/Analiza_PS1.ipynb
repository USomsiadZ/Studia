**Zadanie 1**

Znajdź sumę kolejnych liczb naturalnych od 1 do `N` włącznie.

````python
n = 5
suma = 0 
for i in list(range(0,n)):
    suma += i
print(suma)
````

* drugi element;

* cztery pierwsze elementy;

**Zadanie 3**

Utwórz listę 7-elementową. Następnie w pętli dla każdego elementu wypisz komunikat:
```
Element o indeksie [nr indeksu] jest [parzysty/nieparzysty]!
```

````python
a=[0,1,2,3,4,5,6]
for i in range(len(a)):
    if (i % 2 == 0):
        print(f"{i} jest parzyste")
    else:
        print(f"{i} jest nie parzyste")
````

**Zadanie 4**

Utwórz listę liczb 2-cyfrowych podzielnych przez 5. Następnie wypisz tylko te elementy, które przy dzieleniu przez 3 dają resztę 2.

````python
a = [x for x in range(100) if x % 5 == 0]
print(a)
print([x for x in a if x % 3 == 2])
````
