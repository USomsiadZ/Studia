## **Szeregi**

## Ciąg sum częściowych

Szereg liczbowy to nieskończona suma liczb:

$$S = \sum_{n=1}^{\infty} a_n = a_1 + a_2 + a_3 + \cdots.$$

Mówimy, że szereg jest zbieżny, jeżeli ta suma jest skończona. Wtedy liczbę $S$ nazywamy sumą szeregu.

---

Ciąg sum częściowych to ciąg postaci:

$$s_n = \sum_{k=1}^n a_k = a_1 + \cdots + a_n.$$

Granica ciągu sum częściowych jest równa sumie szeregu, czyli:

$$\lim_{n \to \infty} s_n = S.$$

---

Wyznaczymy ciąg sum częściowych szeregu:

$$S = \sum_{n=1}^\infty \frac{1}{n(n+1)}.$$

```python
var('k')
a(n) = 1/(n*(n+1))

s(n) = sum( a(k), k, 1, n )
print( 'Ciąg sum częściowych:', s(n) )
```

Mamy:

$$s_n = \sum_{k=1}^n \frac{1}{k(k+1)} = \frac{1}{1\cdot(1+1)} + \frac{1}{2\cdot(2+1)} + \frac{1}{3\cdot(3+1)} + \cdots + \frac{1}{n\cdot(n+1)} = \frac{n}{n+1}.$$

---

Następnie liczymy granicę tego ciągu:

```python
suma = lim( s(n), n=oo )
print('Granica ciągu sum częściowych (suma szeregu) jest równa:', suma)
```

Otrzymujemy:

$$\lim_{n \to \infty} s_n = \lim_{n \to \infty} \frac{n}{n+1} = 1.$$

---

Wartość ta jest równa sumie szeregu:

$$S = 1.$$

```python
sum( a(n), n, 1, oo )
```

## Warunek konieczny

**Warunek konieczny zbieżności szeregu**:

Jeżeli szereg $\sum\limits_{n=1}^\infty a_n$ jest zbieżny, to $\lim\limits_{n \to \infty} a_n = 0$.

---

Sprawdzamy, czy warunek konieczny zachodzi dla szeregu:

$$\sum_{n=1}^\infty \frac{1}{n^2}.$$

```python
a(n) = 1/n^2
g = lim( a(n), n=oo )

if g == 0:
    print('Warunek konieczny zbieżności szeregu jest spełniony!')
else:
    print('Warunek konieczny zbieżności szeregu nie jest spełniony.')
```

Oznacza to, że szereg może być zbieżny. Gdyby ten warunek nie zachodził, szereg od razu jest rozbieżny.

---

Znajdujemy sumę szeregu:

```python
suma = sum( a(n), n, 1, oo )
print('Suma szeregu to:', suma)
```

Okazuje się, że:

$$\sum_{n=1}^\infty \frac{1}{n^2} = \frac{\pi^2}{6}.$$

---

Przedstawiamy na rysunku ciąg $(a_n)$ i ciąg sum częściowych $(s_n)$:

```python
rys = points( [(n, a(n)) for n in range(1, 21)], color='green', size=20, legend_label='Ciąg $a_n$', gridlines=True )
rys += points( [(n, sum(a(k), k, 1, n)) for n in range(1, 21)], color='blue', size=20, legend_label='Ciąg sum częściowych $S_n$' )
rys += plot( 0, (x, 0, 20), color='red', linestyle='--', thickness=2, legend_label='Granica ciągu $a_n$' )
rys += plot( suma, (x, 0, 20), color='magenta', linestyle='--', thickness=2, legend_label='Granica ciągu $S_n$' )
rys
```

## Zbieżność bezwzględna

Weźmy szereg zbieżny $\sum\limits_{n=1}^\infty a_n$. Jeżeli szereg modułów $\sum\limits_{n=1}^\infty |a_n|$ jest:

* zbieżny, to szereg jest zbieżny bezwzględnie;
* rozbieżny, to szereg jest zbieżny warunkowo.

---

Sprawdzimy rodzaj zbieżności szeregu:

$$\sum_{n=1}^\infty \frac{(-1)^n}{n}.$$

```python
a(n) = (-1)^n/n
sum( a(n), n, 1, oo )
```

Szereg ten jest zbieżny, ponieważ otrzymujemy skończoną sumę:

$$\sum_{n=1}^\infty \frac{(-1)^n}{n} = -\ln 2.$$

---

Sprawdzamy zbieżność szeregu modułów:

```python
sum( abs( (-1)^n/n ), n, 1, oo )
```

Dostajemy komunikat `Sum is divergent` (szereg jest rozbieżny), więc szereg ten nie jest zbieżny:

$$\sum_{n=1}^\infty \left| \frac{(-1)^n}{n} \right| = \sum_{n=1}^\infty \frac{1}{n} = +\infty.$$

---

Możemy wykorzystać obsługę wyjątków za pomocą składni **try**/**except**, aby wypisać odpowiedni komunikat w przypadku uzyskania tego błędu:

```python
try:
    suma = sum( abs( (-1)^n/n ), n, 1, oo )
    print('Szereg jest bezwzględnie zbieżny!')
except (RuntimeError, ValueError):
    print('Szereg jest warunkowo zbieżny.')
```

## Szereg funkcyjny

Szereg funkcyjny to nieskończona suma funkcji:

$$S(x) = \sum_{n=1}^\infty f_n(x) = f_1(x) + f_2(x) + f_3(x) + \cdots.$$

Dziedzinę tej funkcji nazywamy przedziałem zbieżności szeregu. Wtedy szereg jest funkcją $S$ określoną w tej dziedzinie.

---

Ciąg sum częściowych takiego szeregu jest postaci:

$$s_n(x) = \sum_{k=1}^n f_k(x) = f_1(x) + \cdots + f_n(x).$$

Granica ciągu sum częściowych jest równa sumie $S(x)$ dla każdego $x$ z przedziału zbieżności, czyli:

$$\lim_{n \to \infty} s_n(x) = S(x).$$

---

Wyznaczymy ciąg sum częściowych szeregu:

$$\sum_{n=1}^\infty nx^n.$$

```python
var('k')
f(n) = n * x^n

s(n) = sum( f(k), k, 1, n )

print('Ciąg sum częściowych:')
show( s(n) )
```

Przedział zbieżności tego szeregu funkcyjnego to $(-1, 1)$. Dla każdego $x$ z tego przedziału uzyskany szereg liczbowy jest zbieżny.

---

Liczymy granicę ciągu sum częściowych:

```python
forget()
assume( abs(x) < 1 )  # przedział zbieżności

S(x) = lim( s(n), n=oo )

print('Granica ciągu sum częściowych:')
show(S)
```

Wartość ta dla każdego $x \in (-1, 1)$ jest równa sumie szeregu:

```python
S(x) = sum( f(n), n, 1, oo )

print('Suma szeregu:')
show(S)
```

Przedstawiamy na rysunku tę funkcję, czyli:

$$S \colon\; (-1, 1) \to \mathbb{R}.$$

```python
rys = plot( S, color='red', thickness=3, ymin=-2, ymax=20, gridlines=True, title='Suma szeregu!' )
rys
```

## Szereg potęgowy

Szereg potęgowy to szereg funkcyjny postaci:

$$\sum_{n=0}^\infty a_n (x-x_0)^n.$$

---

Promień zbieżności szeregu potęgowego to liczba:

$$R = \lim_{n \to \infty} \left| \frac{a_n}{a_{n+1}} \right|.$$

---

Wówczas przedział zbieżności takiego szeregu jest postaci:

$$(x_0 - R,\ x_0 + R).$$

Ponadto szereg może być zbieżny na końcach tego przedziału, czyli w punktach $x = x_0 - R$ lub $x = x_0 + R$.

---

Znajdziemy przedział zbieżności szeregu:

$$\sum_{n=0}^\infty \frac{(-1)^n \cdot (x - 3)^n}{(n+1)^2 \cdot 2^n}.$$

Otrzymujemy:

$$a_n = \frac{(-1)^n}{(n+1)^2 \cdot 2^n} \quad \quad \text{oraz} \quad \quad x_0 = 3.$$

Wyznaczamy promień zbieżności:

```python
a(n) = (-1)^n / ((n+1)^2 * 2^n)
x0 = 3

R = lim( abs(a(n)/a(n+1)), n=oo )
print('Promień zbieżności szeregu potęgowego:', R)
```

Zatem szereg jest zbieżny w przedziale:

$$(x_0 - R,\ x_0 + R) = (3 - 2,\ 3 + 2) = (1,\ 5).$$

---

Sprawdzamy zbieżność na końcach tego przedziału:

```python
szereg(x) = a(n) * (x - x0)^n

s1 = sum(szereg(x0-R), n, 0, oo).n()
s2 = sum(szereg(x0+R), n, 0, oo).n()

print( f'Suma szeregu dla x = {x0 - R}:', s1 )
print( f'Suma szeregu dla x = {x0 + R}:', s2 )
```

Po obu stronach suma jest skończona, więc maksymalny przedział zbieżności to:

```python
print(f'Przedział zbieżności szeregu: [{x0 - R}, {x0 + R}]')
```

Liczymy sumę szeregu w tym przedziale zbieżności:

```python
forget()
assume( 1<x, x<5 )  # przedział zbieżności

S(x) = sum( szereg, n, 0, oo )
print( 'Suma szeregu potęgowego:', S(x) )
```

Rysujemy wykres uzyskanej funkcji w przedziale zbieżności:

```python
rys = plot( S, (x, 1, 5), color='red', thickness=2, gridlines=True, title='Wykres szeregu!' )
rys += points( [(1, S(1)), (5, S(5))], color='red', size=30, ymin=-0.5 )
rys
```
