## Wielomiany Taylora

Znajdziemy wielomiany Taylora stopnia 3 i 5 funkcji:

$$f(x) = \sin x.$$

```sage
f(x) = sin(x)

W3 = taylor( f(x), x, 0, 3 )
W5 = taylor( f(x), x, pi, 5 )

print('Wielomian Taylora stopnia 3 o środku w punkcie 0:', W3)
print('Wielomian Taylora stopnia 5 o środku w punkcie pi:', W5)
```

```sage
rys = plot( f, (x, -4, 7), color='red', thickness=3, ymin=-5, ymax=5, title='Wielomiany Taylora!' )
rys += plot( W3, (x, -4, 7), color='blue', legend_label='Wielomian o środku w 0' )
rys += plot( W5, (x, -4, 7), color='green', legend_label=r'Wielomian o środku w $\pi$' )
rys
```

Wyznaczymy kilka początkowych wielomianów funkcji $f$ nieparzystego stopnia:

```sage
for n in range(1, 14, 2):
    print( f'Wielomian stopnia {n}:', taylor(f(x), x, 0, n) )
```

```sage
rys = plot( f, (x, -6, 6), color='red', thickness=3, gridlines=True )
rys += plot( [taylor(f, x, 0, n) for n in range(1, 14, 2)], (x, -6, 6), ymin=-5, ymax=5,
             legend_label=[f'{n = }' for n in range(1, 14, 2)], title='Wielomiany Taylora!' )
rys
```

## Wielomiany trygonometryczne

Znajdziemy wielomian trygonometryczny stopnia 3 funkcji:

$$f(x)=
\left\{
\begin{aligned}
-1, &\quad \text{dla}\; x \in (-2, 0)\\
 0, &\quad \text{dla}\; x \in \{-2, 0, 2\}\\
 1, &\quad \text{dla}\; x \in (0, 2).
\end{aligned}
\right.$$

```sage
f(x) = piecewise( [((-2, 0), -1), ([0, 0], 0), ((0, 2), 1)] )

T3 = f.fourier_series_partial_sum(3)

print('Wielomian trygonometryczny stopnia 7:', T3)
```

```sage
rys = plot( f, (x, -2, 2), color='brown', thickness=2 )
rys += plot( T3, (x, -4, 4), color='forestgreen' )
rys
```

Znajdziemy wielomian trygonometryczny stopnia 5 funkcji:

$$f(x)=
\left\{
\begin{aligned}
 1,        &\quad \text{dla}\; x=-1\\
 3,        &\quad \text{dla}\; x\in(-1,0)\\
 2,        &\quad \text{dla}\; x=0\\
 -2x+1,    &\quad \text{dla}\; x\in(0,1)\\
 1,        &\quad \text{dla}\; x=1.
\end{aligned}
\right.$$

```sage
f(x) = piecewise( [([-1, -1], 1), ((-1, 0), 3), ([0, 0], 2), ((0, 1), -2*x + 1), ([1, 1], 1)] )

T5 = f.fourier_series_partial_sum(5)

print('Wielomian trygonometryczny stopnia 5:')
show( T5 )
```

```sage
rys = plot( f, (x, -1, 1), color='brown', thickness=2 )
rys += plot( T5, (x, -3, 3), color='forestgreen' )
rys
```

## Szereg Fouriera

Szeregiem Fouriera funkcji $f$ w przedziale $[-\ell, \ell]$ nazywamy szereg funkcyjny:

$$\frac{1}{2}a_0 + \sum\limits_{n=1}^{\infty} \left( a_n \cos \frac{n \pi x}{\ell} + b_n \sin \frac{n \pi x}{\ell} \right).$$

---

Wyznaczymy współczynniki $a_7$ i $b_7$ w rozwinięciu funkcji $f(x) = -2x$ w szereg Fouriera w przedziale $[-\pi, \pi]$:

```sage
f(x) = piecewise( [([-pi, pi], -2*x)] )

a7 = f.fourier_series_cosine_coefficient(7)
b7 = f.fourier_series_sine_coefficient(7)

print('Współczynnik a7:', a7)
print('Współczynnik b7:', b7)
```

Wielomiany trygonometryczne to kolejne sumy częściowe szeregu Fouriera. Znajdziemy siódmy składnik nieskończonej sumy:

```sage
s7 = a7 * cos(7*pi*x/pi) + b7 * sin(7*pi*x/pi)
s7
```

```sage
T6 = f.fourier_series_partial_sum(6)
T7 = f.fourier_series_partial_sum(7)
T7 - T6
```

## Transformata Fouriera

Transformatą Fouriera funkcji $f$ nazywamy funkcję:

$$\hat{f}(\omega) = \int\limits_{\mathbb{R}} f(t) e^{-\omega i t} \, dt.$$

---

Znajdziemy transformatę Fouriera dla funkcji $f$:

$$f(t) = \begin{cases}
3, & \text{dla} \ \ t \in (-2, 2) \\
0, & \text{dla} \ \ t \in (-\infty, -2] \cup [2, + \infty).
\end{cases}$$

```sage
var('t w')
f(t) = 3
calka = f(t) * exp(-i*w*t)

# całkujemy w przedziale, gdzie funkcja jest różna od 0
transformata = integrate( calka, (t, -2, 2) )

print('Transformata Fouriera:')
show( transformata )
```
