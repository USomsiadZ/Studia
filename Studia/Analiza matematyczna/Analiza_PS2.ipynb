## Wykres funkcji

```sage
f(x) = sin(x)
plot(f)
```

```sage
# dziedzina
plot( f, (x, -pi, pi) )
```

```sage
# zbiór wyświetlanych wartości
plot( f, (x, -pi, pi), ymin=-2, ymax=2 )
```

```sage
# grubość linii, kolor wykresu, tytuł i legenda
plot( f, (x, -pi, pi), ymin=-2, ymax=2,
      thickness=3, color='green',
      title='Wykres funkcji sinus!',
      legend_label=r'$y=\sin x$' )
```

## Funkcje wymierne

```sage
# wykres jest nieczytelny
f(x) = (2*x+2)/(2*x+1)
plot(f)
```

```sage
# ograniczamy zbiór wartości i usuwamy linię pionową
funkcja = plot( f, (x, -2, 1), ymin=-5, ymax=5, detect_poles=True )
funkcja
```

```sage
asymptota_pionowa = line( [(-0.5, -6), (-0.5, 6)], color='red', linestyle='--' )
asymptota_pozioma = plot( 1, (x, -2, 1), color='green', linestyle='--' )

funkcja + asymptota_pionowa + asymptota_pozioma
```

## Łączenie wykresów

```sage
rys = plot( x^2-4*x+3, (x, -1, 5), thickness=2, title='Wykres funkcji kwadratowej!', gridlines=True )
rys
```

```sage
rys += point( (2, -1), color='red', size=30, zorder=5 )
rys
```

```sage
rys += text( '$W=(2,-1)$', (2.5, -1.3), color='black', fontsize=12 )
rys
```
