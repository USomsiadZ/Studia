## Granica ciągu

```python
a(n) = ( (2*n+6) / (2*n+5) )^n

g = limit( a(n), n=oo )
print('Granica ciągu jest równa:', g)
```

```python
epsilon = 1/4

ciag = points( [(n, a(n)) for n in range(1, 21)], ymin=1, ymax=2, size=10, zorder=3 )
granica = plot( g, (x, 0, 20), color='green', linestyle='--', zorder=2 )
pasek = region_plot( [y > g-epsilon, y < g+epsilon], (x, 0, 20), (y, -1, 3), incol='pink', zorder=1 )

show(ciag + granica + pasek, aspect_ratio=5, figsize=10)
```

**Zadanie 3**

Ciąg $(a_n)$ jest określony dla $n \geq 3$ wzorem:

$$a_n = n - \sqrt{n^2 - 3n}.$$

a) Oblicz granicę ciągu $(a_n)$.

b) Narysuj wykres ciągu $(a_n)$ oraz zaznacz jego granicę linią przerywaną.

```python
a(n) = n - sqrt(n^2 - 3*n)
g = limit(a(n),n = oo)
rys = points( [(n,a(n)) for n in range (3,73)], ymin=1)
rys += plot(g,(n,0,70))
rys
```
