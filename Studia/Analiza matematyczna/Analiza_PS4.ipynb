## **Granice funkcji**

## Wykres funkcji nieciągłej

```sage
# f1(x) dla x < -1
f1(x) = -x - 3

rys = plot( f1, (x, -4, -1), zorder=4 )
rys += point( (-1, -2), color='white', markeredgecolor='blue', size=30, zorder=5 )
rys
```

```sage
# f2(x) dla x >= -1
f2(x) = x^2 - 2*x - 4

rys += plot( f2, (x, -1, 4), zorder=4 )
rys += point( (-1, -1), color='blue', size=30, zorder=5 )
rys
```

## Granica funkcji w nieskończoności

```sage
f(x) = arctan(x)

g1 = lim( f(x), x=-oo )
g2 = lim( f(x), x=oo )

print('Granica przy x dążącym do -oo:', g1)
print('Granica przy x dążącym do +oo:', g2)
```

```sage
rys = plot( f, (x, -10, 10), ymin=-2, ymax=2, color='green', thickness=2 )
rys += plot( g1, (x, -10, 10), color='gray', linestyle='--' )
rys += plot( g2, (x, -10, 10), color='gray', linestyle='--' )
rys
```

## Granica funkcji w punkcie

```sage
f(x) = e^x / x^2

g = lim( f(x), x=0 )

print('Granica przy x dążącym do 0 jest równa:', g)
```

```sage
plot( f, (x, -2, 2), ymin=-1, ymax=21, detect_poles=True, color='green', thickness=2 )
```

## Granice jednostronne

```sage
# f1(x) dla x <= -1, f2(x) dla x > -1
f1(x) = 2*x + 4
f2(x) = x^2 - 2*x - 2

g1 = lim( f1(x), x=-1, dir='-' )
g2 = lim( f2(x), x=-1, dir='+' )

print('Granica lewostronna w punkcie –1 jest równa:', g1)
print('Granica prawostronna w punkcie –1 jest równa:', g2)
```

```sage
rys = plot( f1, (x, -4, -1), zorder=4 )
rys += point( (-1, 2), color='blue', size=30, zorder=5 )
rys += point( (-1, 1), color='white', markeredgecolor='blue', size=30, zorder=5 )
rys += plot( f2, (x, -1, 4), zorder=4 )
rys
```
