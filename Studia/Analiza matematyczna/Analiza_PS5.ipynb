## Asymptota pozioma

```python
# f1 dla x < 1, f2 dla x >= 1
f1(x) = 2*x^2 / (x-1)
f2(x) = (3*x-3) / (x+1)

rys = plot( f1, (x, -5, 1), ymin=-9, ymax=5 )
rys += plot( f2, (x, 1, 11) )
rys += point( (1, f2(1)), size=20, zorder=5 )
rys
```

```python
g1 = lim( f1(x), x=-oo )
g2 = lim( f2(x), x=oo )

print('Granica przy x dążącym do -oo:', g1)
print('Granica przy x dążącym do +oo:', g2)
```

```python
asymptota_pozioma = plot( 3, (x, -5, 11), color='green', linestyle='--',
                          legend_label='Asymptota pozioma prawostronna $y=3$' )
rys + asymptota_pozioma
```

## Asymptota pionowa

```python
g1 = lim( f1(x), x=1, dir='-' )
g2 = lim( f2(x), x=1, dir='+' )

print(f'Granica lewostronna w punkcie 1 jest równa:', g1)
print(f'Granica prawostronna w punkcie 1 jest równa:', g2)
```

```python
asymptota_pionowa = line( [(1, -10), (1, 6)], color='red', linestyle='--',
                          legend_label='Asymptota pionowa lewostronna $x=1$' )
asymptota_pozioma = plot( 3, (x, -5, 11), color='green', linestyle='--',
                          legend_label='Asymptota pozioma prawostronna $y=3$' )
rys += asymptota_pozioma`~
rys += asymptota_pionowa
rys
```

```python
a = lim( f1(x)/x, x=-oo )
print('Współczynnik kierunkowy:', a)

b = lim( f1(x)-a*x, x=-oo )
print('Wyraz wolny:', b)
```
