Dana jest funkcja:
\[
f(x)=
\begin{cases}
\dfrac{5-3x-x^2-x^3}{x^2-4}, & x<-2,\\[6pt]
2, & x=-2,\\[6pt]
\dfrac{3x-1}{x+2}, & x>-2.
\end{cases}
\]

a) Oblicz granice jednostronne funkcji \(f\) w punkcie \(x_0=-2\).

b) Wyznacz wszystkie asymptoty wykresu funkcji \(f\) oraz podaj ich rodzaj.

c) Narysuj wykres funkcji \(f\); zaznacz na rysunku asymptoty innym kolorem i linią przerywaną.

---

```python
f1(x) = (5-3*x-x^2-x^3) / (x^2-4)
f2(x) = (3*x-1) / (x+2)
g1 = lim(f1(x),x=-2,dir='-')
g2 = lim(f2(x),x=-2,dir='+')
print(g1,g2)
```

```python
g1 = lim(f1(x),x=-oo)
g2 = lim(f2(x),x=oo)
print(g1)
print(g2)
```

```python
rys = plot(f1,(x,-5,-2),ymin=-10,ymax=10)
rys += plot(f2,(x,-2,5))
rys += point((-2,2),size=10)
rys += plot(3,(-5,5),color='grey',linestyle='--')
rys += line([(-2,-10),(-2,10)],color='red',linestyle='--')
a = lim(f1(x)/x,x=-oo)
b = lim(f1(x)-a*x,x=-oo)

print(a,b)
rys += plot(a*x+b,(x,-5,5),color='green',linestyle='--')
rys
```
