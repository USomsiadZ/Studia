### Widać enum w debugerze
```c

enum kolor {
    RED,
    GREEN,
    BLUE
};
enum kolor aktualny_kolor = RED;
// Enum -> int
void ustaw_aktualny(enum kolor k){
    aktualny_kolor = k;
}
int main() {
    ustaw_aktualny(2); // BLUE
    printf("%d", aktualny_kolor);
    return 0;
}
```

### Praca z plikami
- fopen() — otwiera plik
- fprintf() / fputs() / fputc() / fwrite() — zapis
- fscanf() / fgets() /  fgetc() / fread() — odczyt
- fclose() — zamyka plik
- fseek(p, 5, SEEK_SET)
- "r" — tylko czytanie
- "w" — zapis (czyści plik)
- "a" — dopisanie
- "rb", "wb", "ab" — wersje binarne
- "r+", "w+", "a+" — czytanie + zapis

````c
#include <stdio.h>
int main(){
    char buf[100];
    FILE *p = fopen("dane.txt", "r+");
    //Dopisanie
    fprintf(p, "Hello!");
    
    //Odczytanie
    while (fgets(buf, sizeof(buf), p) != NULL) {
        printf("%s", buf);
    }
    
    fclose(p);
}
````

- SEEK_SET od początku pliku 

- SEEK_CUR od aktualnej pozycji

- SEEK_END od końca pliku





````c
#include <stdio.h>
int main(){
    FILE *plik = fopen("kordynaty.txt", "w+");
    fseek(plik, 2,SEEK_SET);
    fputc('0', plik);
    fputc('1', plik);
    fputc('2', plik);
    fclose(plik);

}
````

````c
// Adresy
#include <stdio.h>
int main(){
    const char *zdanie = "Witaj Świecie"; // wskazuje na W
    printf("%c\n",zdanie[0]);
    printf("%s\n",zdanie);
}
````

````c
#include <stdio.h>
// &a Jaki jest adres zmiennej w pamieci
// *a Co sie znajduje pod zmienna w pamieci
// printf %p do wypisywania adresu
int main() {
    int a = 1;      // normalna zmienna
    int *x = &a;
    
    printf("%p\n", &a); // wypisz adres
    printf("%d\n", *x); // wypisz adres

    return 0;
}
````
