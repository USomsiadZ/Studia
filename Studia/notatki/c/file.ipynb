# Obsługa plików w C

## Podstawowe operacje na plikach

W języku C do pracy z plikami używamy funkcji z biblioteki `<stdio.h>`.

### Kluczowe elementy:

| Funkcja | Opis | Przykład |
|---------|------|----------|
| `fopen()` | Otwiera plik | `FILE *f = fopen("dane.txt", "r");` |
| `fclose()` | Zamyka plik | `fclose(f);` |
| `fprintf()` | Zapisuje sformatowane dane do pliku | `fprintf(f, "Liczba: %d", 42);` |
| `fscanf()` | Czyta sformatowane dane z pliku | `fscanf(f, "%d", &x);` |
| `fread()` | Czyta blok danych binarnych | `fread(buf, sizeof(int), 10, f);` |
| `fwrite()` | Zapisuje blok danych binarnych | `fwrite(buf, sizeof(int), 10, f);` |

### Tryby otwierania plików:

| Tryb | Opis | Przykład użycia |
|------|------|-----------------|
| `"r"` | Odczyt (plik musi istnieć) | `fopen("config.txt", "r")` |
| `"w"` | Zapis (tworzy nowy lub nadpisuje) | `fopen("log.txt", "w")` |
| `"a"` | Dopisywanie na końcu | `fopen("historia.txt", "a")` |
| `"r+"` | Odczyt i zapis (plik musi istnieć) | `fopen("baza.txt", "r+")` |
| `"w+"` | Odczyt i zapis (tworzy nowy lub nadpisuje) | `fopen("temp.txt", "w+")` |
| `"rb"` | Odczyt binarny | `fopen("obraz.bin", "rb")` |
| `"wb"` | Zapis binarny | `fopen("dane.bin", "wb")` |

---

## Przykład: Otwieranie i zamykanie pliku

````c
FILE *f = fopen("sensor.log", "r");
if (!f) {
    fprintf(stderr, "Error opening sensor.log\n");
    return 1;
}
fclose(f);
````

````c
// Czytanie linii z pliku
char buffer[256];
while (fgets(buffer, sizeof(buffer), f) != NULL) {
    printf("%s", buffer);
}

// Czytanie sformatowanych danych
int value;
fscanf(f, "%d", &value);

// Zapisywanie do pliku
FILE *out = fopen("output.txt", "w");
fprintf(out, "Wartość: %d\n", value);
fclose(out);
````

## Operacje binarne na plikach

```c
// Zapis binarny
int dane[] = {1, 2, 3, 4, 5};
FILE *f = fopen("dane.bin", "wb");
fwrite(dane, sizeof(int), 5, f);  // zapisuje 5 intów
fclose(f);

// Odczyt binarny
int odczytane[5];
FILE *f2 = fopen("dane.bin", "rb");
fread(odczytane, sizeof(int), 5, f2);  // czyta 5 intów
fclose(f2);
```

## Nawigacja w pliku

| Funkcja | Opis | Przykład |
|---------|------|----------|
| `fseek(f, offset, origin)` | Przesuwa wskaźnik pliku | `fseek(f, 100, SEEK_SET);` — idź do bajtu 100 |
| `ftell(f)` | Zwraca aktualną pozycję | `long pos = ftell(f);` |
| `rewind(f)` | Wraca na początek pliku | `rewind(f);` — równoważne `fseek(f, 0, SEEK_SET)` |

**Origin w `fseek()`:**

| Stała | Opis | Przykład |
|-------|------|----------|
| `SEEK_SET` | Od początku pliku | `fseek(f, 0, SEEK_SET);` — początek |
| `SEEK_CUR` | Od aktualnej pozycji | `fseek(f, -5, SEEK_CUR);` — 5 bajtów wstecz |
| `SEEK_END` | Od końca pliku | `fseek(f, 0, SEEK_END);` — koniec pliku |
