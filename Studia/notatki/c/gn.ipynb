### Instrukcja continue

- Powoduje **natychmiastowe przejście do sprawdzenia warunku przed kolejną iteracją**
- Pomija pozostały kod w bieżącej iteracji
- Działa w każdej pętli (for, while, do...while)
- Nie kończy pętli, tylko bieżącą iterację – w przeciwieństwie do break

**Przykład:**
```c
#include <stdio.h>
int main(void) {
    int number, count = 0, attempts = 0;
    while (count < 5 && attempts < 10) {
        printf("Number %d: ", count + 1);
        scanf("%d", &number);
        attempts++;
        if (number <= 0) {
            printf("Invalid input (%d/10).\n", attempts);
            continue;
        }
        printf("Accepted: %d\n", number);
        count++;
    }
    return 0;
}
```

## Tablice napisów

W C reprezentowane są jako tablice dwuwymiarowe znaków:
- pojedynczy napis: `char name[] = "Data"` ( cztery znaki plus znak końca `'\0'`)
- tablica napisów to więc tablica tablic znaków: `char crew[5][20]` (pięć nazwisk po 19 znaków każdy)


**Dostęp do elementów:**
- Cały napis (nazwisko): `crew[1]` → "Riker"
- Pojedynczy znak w danym nazwisku: `crew[2][0]` → 'D' w "Data"

**Wczytywanie napisów do tablicy:**

```c
for (int i = 0; i < 5; i++) {

}
```

## Organizacja projektu wieloplikowego

### Pliki nagłówkowe (.h) i implementacyjne (.c)

**Plik nagłówkowy (starfleet_math.h):**

```c
#ifndef STARFLEET_MATH_H
#define STARFLEET_MATH_H

double warp_energy(double warp_factor);

#endif
```

**Plik implementacyjny (starfleet_math.c):**

```c
#include "starfleet_math.h"

double warp_energy(double warp_factor) {
    return warp_factor * warp_factor * warp_factor;
}
```

**Plik główny (main.c):**

```c
#include <stdio.h>
#include "starfleet_math.h"

int main() {
    double warp = 5.0;
    double energy = warp_energy(warp);
    
    printf("Ustawiony kurs: warp %.1f\n", warp);
    printf("Zapotrzebowanie na energię: %.2f jednostek\n", energy);
    
    return 0;
}
```

### Kompilacja projektu wieloplikowego

Kompilacja programu wieloplikowego odbywa się w dwóch krokach:

1. Każdy plik .c jest tłumaczony osobno do pliku obiektowego .o
2. Linker łączy wszystkie .o w jeden plik wykonywalny

````python
# Krok 1: Kompilacja do plików obiektowych
gcc -c starfleet_math.c    # Tworzy starfleet_math.o
gcc -c main.c              # Tworzy main.o

# Krok 2: Linkowanie
gcc main.o starfleet_math.o -o enterprise
````

## Makefile

### Podstawy Makefile

Makefile automatyzuje proces kompilacji i pozwala na szybkie budowanie projektów.

````python
# Kompilator i opcje
CC = gcc
CFLAGS = -Wall

# Domyślny cel
all: enterprise

# Linkowanie programu
enterprise: main.o starfleet_math.o
	$(CC) main.o starfleet_math.o -o enterprise

# Kompilacja modułów
main.o: main.c starfleet_math.h
	$(CC) $(CFLAGS) -c main.c

starfleet_math.o: starfleet_math.c starfleet_math.h
	$(CC) $(CFLAGS) -c starfleet_math.c

# Uruchomienie programu
run: all
	./enterprise

# Sprzątanie
clean:
	rm -f *.o enterprise
````

### Uruchamianie Makefile

```bash
make          # Buduje projekt (wykonuje cel 'all')
make run      # Kompiluje i uruchamia program
make clean    # Usuwa pliki obiektowe i plik wykonywalny
```

### Wyjaśnienia Makefile

**CC, CFLAGS:**
- `CC = gcc` – zmienna z nazwą kompilatora (np. gcc, clang)
- `CFLAGS = -Wall` – opcje przekazywane podczas kompilacji plików .c do .o
- **Po co?** Ustawia kompilator/opcje dla całego projektu – nie trzeba powtarzać „ręcznie" flag w każdej regule

**all – cel domyślny:**
- `all: enterprise`
- „Główny" cel, który buduje cały program/projekt
- Gdy wpiszesz samo `make`, wykonywany jest pierwszy cel w pliku – warto, żeby był nim `all`
- **Po co?** Porządek i przewidywalność: zawsze wiesz, że `make` zbuduje projekt „jak trzeba"

**run – uruchamianie programu z Makefile:**
- `run: all`
- `./enterprise`
- Wygodny cel, który najpierw buduje program (jeśli potrzeba), a potem go uruchamia
- **Po co?** Jedno polecenie do kompilacji i startu – idealne na ćwiczenia i szybkie testy

**clean – sprzątanie:**
- `clean:`
- `rm -f *.o enterprise`
- Usuwa pliki obiektowe i plik wykonywalny
- **Po co?** Pozwala „odświeżyć" projekt – usuwa stare pliki, czyści po błędach

## Zmienne globalne – minimalizacja ryzyka

#### Ukrywaj zmienne w module (static)

Użyj `static` przy zmiennej globalnej, jeśli nie musi być widoczna w innych plikach. Dzięki temu nikt spoza pliku nie będzie mógł jej przypadkowo zmienić.

````python
static int reactorEnergy = 100;
````

#### Używaj const dla wartości niezmiennych

Jeśli coś ma być tylko do odczytu, zadeklaruj to jako `const`. Zapobiegnie to przypadkowemu nadpisaniu danych.

````python
const double MAX_WARP_SPEED = 9.9;
````

#### Korzystaj z dyrektywy #define

Stosuj `#define` dla globalnych stałych.

````python
#define MAX_WARP_SPEED 9.9
````

#### Stosuj spójne nazwy dla globalnych danych

Wprowadź konwencję, np. prefiks `g_` lub `global_`, aby od razu było widać, że zmienna jest globalna.

````python
int g_shieldStatus = 1;
````

#### Używaj tylko wtedy, gdy musisz

Jeśli to możliwe, pisz funkcje tak, by nie wymagały globalnych danych – przekazuj im stan przez parametry.

#### Dokumentacja

Każda nowa globalna powinna mieć uzasadnienie i być wpisana do dokumentacji. Rób okresowe „code review" globalnych zmiennych.
