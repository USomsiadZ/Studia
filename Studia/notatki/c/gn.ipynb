
### Instrukcja continue

- Powoduje **natychmiastowe przejście do sprawdzenia warunku przed kolejną iteracją**
- Pomija pozostały kod w bieżącej iteracji
- Działa w każdej pętli (for, while, do...while)
- Nie kończy pętli, tylko bieżącą iterację – w przeciwieństwie do break

**Przykład:**
```c
#include <stdio.h>

int main(void) {
    for (int i = 1; i <= 5; i++) {
        if (i == 3) {
            continue;  // Pomija liczbę 3
        }
        printf("%d ", i);
    }
    return 0;
}
```
