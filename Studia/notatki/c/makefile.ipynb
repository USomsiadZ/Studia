## Makefile

### Podstawy Makefile

Makefile automatyzuje proces kompilacji i pozwala na szybkie budowanie projektów.


# Kompilator i opcje
CC = gcc
CFLAGS = -Wall

# Domyślny cel
all: enterprise

# Linkowanie programu
enterprise: main.o starfleet_math.o
	$(CC) main.o starfleet_math.o -o enterprise

# Kompilacja modułów
main.o: main.c starfleet_math.h
	$(CC) $(CFLAGS) -c main.c

starfleet_math.o: starfleet_math.c starfleet_math.h
	$(CC) $(CFLAGS) -c starfleet_math.c

# Uruchomienie programu
run: all
	./enterprise

# Sprzątanie
clean:
	rm -f *.o enterprise


### Uruchamianie Makefile

```bash
make          # Buduje projekt (wykonuje cel 'all')
make run      # Kompiluje i uruchamia program
make clean    # Usuwa pliki obiektowe i plik wykonywalny
```


### Wyjaśnienia Makefile

**CC, CFLAGS:**
- `CC = gcc` – zmienna z nazwą kompilatora (np. gcc, clang)
- `CFLAGS = -Wall` – opcje przekazywane podczas kompilacji plików .c do .o
- **Po co?** Ustawia kompilator/opcje dla całego projektu – nie trzeba powtarzać „ręcznie" flag w każdej regule

**all – cel domyślny:**
- `all: enterprise`
- „Główny" cel, który buduje cały program/projekt
- Gdy wpiszesz samo `make`, wykonywany jest pierwszy cel w pliku – warto, żeby był nim `all`
- **Po co?** Porządek i przewidywalność: zawsze wiesz, że `make` zbuduje projekt „jak trzeba"

**run – uruchamianie programu z Makefile:**
- `run: all`
- `./enterprise`
- Wygodny cel, który najpierw buduje program (jeśli potrzeba), a potem go uruchamia
- **Po co?** Jedno polecenie do kompilacji i startu – idealne na ćwiczenia i szybkie testy

**clean – sprzątanie:**
- `clean:`
- `rm -f *.o enterprise`
- Usuwa pliki obiektowe i plik wykonywalny
- **Po co?** Pozwala „odświeżyć" projekt – usuwa stare pliki, czyści po błędach
