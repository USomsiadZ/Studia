

## Organizacja projektu wieloplikowego
**Plik nagłówkowy (starfleet_math.h):**

```c
#ifndef STARFLEET_MATH_H
#define STARFLEET_MATH_H

double warp_energy(double warp_factor);

#endif
```

**Plik implementacyjny (starfleet_math.c):**

```c
#include "starfleet_math.h"

double warp_energy(double warp_factor) {
    return warp_factor * warp_factor * warp_factor;
}
```

**Plik główny (main.c):**

```c
#include <stdio.h>
#include "starfleet_math.h"

int main() {
    double warp = 5.0;
    double energy = warp_energy(warp);
    
    printf("Ustawiony kurs: warp %.1f\n", warp);
    printf("Zapotrzebowanie na energię: %.2f jednostek\n", energy);
    
    return 0;
}
```
