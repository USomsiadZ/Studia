# register
```c
register int i;
```
* Sugeruje kompilatorowi, że dla szybszego dostępu zmienna powinna być przechowywana w
rejestrze procesora zamiast w pamięci RAM.
* Dla takich zmiennych nie można pobrać adresu (&).
# static
```c
static int shieldStatus = 1;
```
* Zachowuje wartość zmiennej między wywołaniami funkcji (dla zmiennych lokalnych).
* Ogranicza widoczność zmiennej lub funkcji do pliku, w którym jest zdefiniowana (dla
elementów globalnych).
* Używany często do ukrywania danych lub pamiętania stanu funkcji.
```c
void licz_wywolania() {
    // Ta zmienna jest statyczna, co oznacza, że:
    // 1. Zostanie zainicjalizowana TYLKO RAZ (przy pierwszym wywołaniu).
    // 2. Zachowa swoją wartość po zakończeniu działania funkcji.
    static int licznik = 0; 
    
    licznik++;
    printf("Wywołanie numer: %d\n", licznik);
}
```
# extern
```c 
extern int reactorEnergy;
```
* Deklaruje zmienną lub funkcję, która jest zdefiniowana w innym pliku źródłowym.
* Umożliwia współdzielenie danych między modułami programu.
* Nie tworzy zmiennej – jedynie wskazuje, że istnieje gdzieś indziej.
```c 
// ship.c
const char *SHIP_NAME = "USS Enterprise NCC-1701-D";
// diagnostics.c
#include "ship.c"
extern const char *SHIP_NAME;
```

## Przekazywanie do funkcji
```c
void move(struct Point *p, int dx, int dy) {
    p->x += dx;
    p->y += dy;
}

void print_point(struct Point p) {     // kopia całej struktury
    printf("(%d, %d)\n", p.x, p.y);
}
```
- Duże struktury przekazuj przez wskaźnik, aby uniknąć kosztownej kopii.

## `sizeof`, wyrównanie (alignment) i padding
```c
struct A { char c; int i; }; // często ma padding między c i i
struct B { int i; char c; }; // inny układ i potencjalnie inny sizeof

printf("sizeof(A)=%zu sizeof(B)=%zu\n", sizeof(struct A), sizeof(struct B));
```
