### Substytucje i zmienne – skrót z prezentacji

| Temat | Najważniejsze | Przykład |
|-------|---------------|----------|
| Rodzaje substytucji | Nazwy plików, poleceń, zmiennych | `echo $(pwd)`; `echo $HOME` |
| Globbing (nazwy plików) | `*` dowolny ciąg; `?` jeden znak; `[a-z]` zakres; `[!0-9]` negacja | `ls a*.txt`; `ls a?.txt`; `ls a[0-9]` |
| Substytucja poleceń | Wynik polecenia wstawiany do innego | `` echo `pwd` `` lub `echo $(pwd)` |
| Substytucja zmiennych | Ustawienie/odczyt przez `nazwa=wartość`, odczyt `echo $nazwa` | `x=41; echo $x`; `unset x` |
| Ograniczniki / cudzysłowy | `""` pozwala na rozwinięcie zmiennych; `''` blokuje; backtick/`$()` dla poleceń | `echo "$HOME"`; `echo 'a $HOME'` |
| PATH | Lista katalogów przeszukiwanych dla poleceń | `echo $PATH`; `PATH=$PATH:/home/u/bin` |
| Export | Udostępnia zmienną procesom potomnym | `A=AAAA; export A; ./script` |
| Podgląd zmiennych | `printenv` (wyeksportowane), `set` (wszystkie), `unset` (kasuje) | `printenv | grep AA`; `unset AA` |
| Dziedziczenie w podpowłoce | Zmiany w podpowłoce nie wracają do macierzystej | `bash; A=BBB; exit` → w rodzicu `A` bez zmian |
| unset | Usuwa zmienną środowiskową|
| a=* | ustawia zmienna srodowiskową|
| export a| exportuje zmienną |


| Polecenie | Opis | Przykład użycia | Przykładowe parametry/opcje |
|-----------|------|-----------------|----------------------------|
| type | Informacja o umiejscowieniu polecenia (typ polecenia). | `type ls` | `-a` (wyświetl wszystkie lokalizacje) |
| whereis | Informacja o plikach binarnych, źródłowych i man. | `whereis passwd` | |


### wc - Word Count (liczenie linii, słów, bajtów)

| Parametr | Opis | Przykład | Wynik |
|----------|------|----------|-------|
| `wc -l` | Tylko liczba linii | `wc -l plik.txt` | `10 plik.txt` |
| `wc -w` | Tylko liczba słów | `wc -w plik.txt` | `50 plik.txt` |
| `wc -m` | Tylko liczba znaków | `wc -m plik.txt` | `295 plik.txt` |
| `wc *.txt` | Dla wielu plików (suma na końcu) | `wc *.txt` | Wyświetla dla każdego + total |

### Przekierowania

| Symbol | Funkcja | Opis | Przykład |
|--------|---------|------|----------|
| `< plik` | Przekierowanie stdin | Stdin polecenia jest przekierowywane z pliku | `wc -l < plik.txt`; `sort < dane.txt` |
| `<< EOF` | Here Document | Umożliwia skopiowanie wieloliniowego tekstu do stdin, aż do napotkania znacznika (np. EOF) | `cat << EOF`<br>`linia 1`<br>`linia 2`<br>`EOF` |
| `2> plik` | Przekierowanie stderr | Przekierowuje błędy do pliku | `ls nieistniejący 2> error.log` |
| `&> plik` | Przekierowanie stdout i stderr | Przekierowuje wszystko do jednego pliku | `./skrypt.sh &> output.log` |
