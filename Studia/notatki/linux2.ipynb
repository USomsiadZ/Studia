### awk - przetwarzanie i analiza tekstu

| Parametr/Składnia | Opis | Przykład | Wynik |
|-------------------|------|----------|-------|
| `awk '{print}'` | Wypisuje całą linię | `echo "a b c" \| awk '{print}'` | `a b c` |
| `awk '{print $1}'` | Wypisuje pierwszą kolumnę | `echo "a b c" \| awk '{print $1}'` | `a` |
| `awk '{print $2}'` | Wypisuje drugą kolumnę | `echo "a b c" \| awk '{print $2}'` | `b` |
| `awk '{print $NF}'` | Wypisuje ostatnią kolumnę | `echo "a b c" \| awk '{print $NF}'` | `c` |
| `awk '{print $1, $3}'` | Wypisuje kolumny 1 i 3 | `echo "a b c d" \| awk '{print $1, $3}'` | `a c` |
| `awk -F:` | Zmienia separator (delimiter) | `awk -F: '{print $1}' /etc/passwd` | Nazwy użytkowników |
| `awk 'NR==3'` | Wypisuje linię numer 3 | `awk 'NR==3' plik.txt` | Trzecia linia |
| `awk 'NR>5'` | Wypisuje linie od 6 wzwyż | `awk 'NR>5' plik.txt` | Linie 6, 7, 8... |
| `awk '/wzorzec/'` | Filtruje linie zawierające wzorzec | `awk '/ERROR/' log.txt` | Linie z "ERROR" |
| `awk '$3 > 100'` | Filtruje gdzie kolumna 3 > 100 | `awk '$3 > 100' dane.txt` | Linie spełniające warunek |
| `awk '{sum+=$1} END {print sum}'` | Sumuje pierwszą kolumnę | `awk '{sum+=$1} END {print sum}' liczby.txt` | Suma |
| `awk 'BEGIN {print "Start"}'` | Wykonuje przed przetwarzaniem | `awk 'BEGIN {print "Kolumna 1"} {print $1}'` | Nagłówek + dane |
| `awk 'END {print "Koniec"}'` | Wykonuje po przetworzeniu | `awk '{print} END {print "---"}' plik.txt` | Dane + stopka |

#### Zmienne wbudowane:
| Zmienna | Opis | Przykład |
|---------|------|----------|
| `$0` | Cała linia | `awk '{print $0}'` |
| `$1, $2, ...` | Kolumny 1, 2, ... | `awk '{print $1, $2}'` |
| `$NF` | Ostatnia kolumna | `awk '{print $NF}'` |
| `NR` | Numer bieżącej linii | `awk '{print NR, $0}'` |
| `NF` | Liczba kolumn w linii | `awk '{print NF}'` |
| `FS` | Separator pól (domyślnie spacja) | `awk 'BEGIN {FS=":"} {print $1}'` |

#### Praktyczne przykłady:
```bash
# Wypisz drugą kolumnę z ls -l
ls -l | awk '{print $2}'

# Suma rozmiaru plików
ls -l | awk '{sum+=$5} END {print sum}'

# Pierwsze pole z pliku CSV
awk -F, '{print $1}' dane.csv

# Filtruj i formatuj
awk '$3 > 50 {print $1, "ma", $3}' dane.txt

# Wypisz linie od 10 do 20
awk 'NR>=10 && NR<=20' plik.txt

# Policz linie zawierające "ERROR"
awk '/ERROR/ {count++} END {print count}' log.txt

# Średnia z pierwszej kolumny
awk '{sum+=$1; count++} END {print sum/count}' liczby.txt

# Usuń puste linie
awk 'NF > 0' plik.txt

# Wypisz tylko unikalne linie
awk '!seen[$0]++' plik.txt
```

## Podstawowe komendy grep

| Komenda | Opis | Przykład | Input | Output |
|---------|------|----------|-------|--------|
| `grep` | Wyszukuje linie pasujące do wzorca | `grep "error" log.txt` | `info msg`<br>`error found`<br>`warning` | `error found` |
| `grep -i` | Ignoruj wielkość liter | `grep -i "error" log.txt` | `Error found`<br>`info msg`<br>`ERROR here` | `Error found`<br>`ERROR here` |
| `grep -v` | Linie NIE pasujące | `grep -v "error" log.txt` | `info msg`<br>`error found`<br>`ok` | `info msg`<br>`ok` |
| `grep -n` | Wyświetla numery linii | `grep -n "error" log.txt` | `info msg`<br>`error found`<br>`ok` | `2:error found` |
| `grep -c` | Zlicza pasujące linie | `grep -c "error" log.txt` | `error 1`<br>`error 2`<br>`ok` | `2` |
| `grep -r` | Przeszukiwanie rekurencyjne | `grep -r "TODO" ./` | `./file1.txt: TODO fix`<br>`./file2.txt: done` | `./file1.txt: TODO fix` |
| `grep -E` | Rozszerzone regex (ERE) | `grep -E "cat\|dog" file.txt` | `I have a cat`<br>`bird`<br>`dog barks` | `I have a cat`<br>`dog barks` |

## Metaznaki w wyrażeniach regularnych

| Metaznak | Opis | Przykład | Input | Output |
|----------|------|----------|-------|--------|
| `.` | Dowolny pojedynczy znak | `grep "a.c"` | `abc`<br>`adc`<br>`ac`<br>`a12c` | `abc`<br>`adc`<br>`a12c` |
| `*` | Zero lub więcej poprzedniego | `grep "ab*c"` | `ac`<br>`abc`<br>`abbc`<br>`ad` | `ac`<br>`abc`<br>`abbc` |
| `^` | Początek linii | `grep "^Jan"` | `Jan kowalski`<br>`Ala Jan`<br>`Jan` | `Jan kowalski`<br>`Jan` |
| `$` | Koniec linii | `grep "end$"` | `the end`<br>`end here`<br>`end` | `the end`<br>`end` |
| `[]` | Klasa znaków | `grep "[abc]"` | `apple`<br>`dog`<br>`cat` | `apple`<br>`cat` |
| `[^]` | Negacja klasy | `grep "[^0-9]"` | `abc`<br>`123`<br>`a1` | `abc`<br>`a1` |
| `\` | Znak ucieczki | `grep "\."` | `file.txt`<br>`file`<br>`3.14` | `file.txt`<br>`3.14` |
| `+` | Jeden lub więcej (ERE) | `grep -E "ab+c"` | `ac`<br>`abc`<br>`abbc` | `abc`<br>`abbc` |
| `?` | Zero lub jeden (ERE) | `grep -E "colou?r"` | `color`<br>`colour`<br>`colouur` | `color`<br>`colour` |
| `\|` | Alternatywa (ERE) | `grep -E "cat\|dog"` | `cat meow`<br>`bird`<br>`dog bark` | `cat meow`<br>`dog bark` |
| `()` | Grupowanie | `grep -E "(ab)+"` | `ab`<br>`abab`<br>`ba` | `ab`<br>`abab` |
| `{n}` | Dokładnie n wystąpień | `grep -E "a{3}"` | `aa`<br>`aaa`<br>`aaaa` | `aaa`<br>`aaaa` |
| `{n,}` | Co najmniej n | `grep -E "a{2,}"` | `a`<br>`aa`<br>`aaa` | `aa`<br>`aaa` |
| `{n,m}` | Od n do m | `grep -E "a{2,3}"` | `a`<br>`aa`<br>`aaa`<br>`aaaa` | `aa`<br>`aaa` |
