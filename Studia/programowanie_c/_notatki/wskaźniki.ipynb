### <b>*p</b> - wartość zmiennej z adresu
### <b>&p</b> - adres zmiennej z wartosci

| Deklaracja | Mogę zmienić `*p` (dane)? | Mogę zmienić `p` (adres)? |
| :--- | :---: | :---: |
| `int *p` | ✅ TAK | ✅ TAK |
| `const int *p` | ❌ NIE | ✅ TAK |
| `int * const p` | ✅ TAK | ❌ NIE |
| `const int * const p` | ❌ NIE | ❌ NIE |

Wskaźnik adresu dowolnego typu
```c
void *p;
```

#### (*p).x lub p->x daje wartość **x**
```c
struct Galaktyka{
    int x;
};
struct Galaktyka *p = &star;

(*p).x
p->x
```

````c
#include <stdio.h>
#include <stdlib.h>
int main(){
    int a = 1;
    int *b = &a;
    int c = 5;
    printf("%p\n",&a);
    printf("%p\n",b);
    *b += 1;
    printf("%d\n",*b);
    printf("%d\n",a);
    b = &c;
    printf("%d\n",*b);


    return 0;
}
````

````c
#include <stdio.h>
#include <stdlib.h>
//*(*(tab + i) + j) dla tab[2][2]
int main(){
    int tab[6] = {0,1,2,3,4,5};
    int *p = &tab[0];
    printf("Wartosc: %d\n",*(p + 1));
    int roznica_w_pamiec = &tab[5] - &tab[1];
    printf("Roz. w pamieci %d\n",roznica_w_pamiec);
    printf("Wartosc tablicy %d\n",p[1]);
    printf("Wartosc tablicy %d\n",*(p+1));
    printf("Wartosc tablicy %d\n",*(tab + 1));
    return 0;
}
````

````c
#include <stdio.h>
#include <stdlib.h>
struct Galaktyka{
    int x;
    int y;
    int z;
};
int main(){
    struct Galaktyka star = {10,12,3000};
    struct Galaktyka *p = &star;
    printf("%d\n",(*p).x);
    printf("%d\n",p->x);
    scanf("%d",&p->x);
    printf("%d\n",p->x);
    return 0;
}
````
