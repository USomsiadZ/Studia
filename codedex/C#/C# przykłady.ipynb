````polyglot-notebook
//Method overload
class Program
{
    public static void Metoda(string pr = "test", string pr2 = "2")
    {
        Console.WriteLine(pr + pr2);
    }
    public static void Metoda(string pr = "test", int pr2 = 2)
    {
        Console.WriteLine(pr + pr2);
    }
}
Program.Metoda(pr2: "3");
Program.Metoda(pr2: 3);
````

````polyglot-notebook
//Automatic property 
class Rob()
{
    // public string Name { get; private set; }. ↔ aktualny zapis
    private string name;
    public string Name
    {
        get { return "liczba: " + name; }
        set { name = value; }
    }
}
class Program
{
    static void Main()
    {
        Rob p = new()
        {
            Name = "1"
        };
        Console.WriteLine(p.Name);

    }
}
````

````polyglot-notebook
//Inheritance
class Rob(){
    protected string name1 = "Ford";
}
class Car : Rob
{

    public string name2 = "Mustang";
    // new służy temu aby pokazac kompilatorowi że świadomie uzywamy protected name1
    public new string name1
    {
        get{return base.name1;}
    }
}
Car car = new();
Console.WriteLine(car.name1);
Console.WriteLine(car.name2);
````

````polyglot-notebook
//Polymorphism
// Virtual -- override - Przez to można nadpisywać metoda
// abstract - Nie można utworzyć obiektu z klasy abstrakcyjnej
// abstract method - obietnica, że w klasie pochodnej będzie implementacja
// sealed - Nie można dziedziczyć z klasy


abstract class Animal
{
  public virtual void animalSound() 
  {
    Console.WriteLine("The animal makes a sound");
  }
  abstract public void animalSound2(); // body nie da się
}

class Pig : Animal 
{
  public override void animalSound() 
  {
    Console.WriteLine("The pig says: wee wee");
  }
  public override void animalSound2()
  {
    Console.WriteLine("The pig says: wee wee weee");
  }
}

//Animal myAnimal = new Animal(); Po dodaniu abstract przestało działał
Animal myPig = new Pig();

//myAnimal.animalSound();
myPig.animalSound();
myPig.animalSound2();
````

````polyglot-notebook
// interface - obietnica, że w klasie pochodnej będzie implementacja
interface IAnimal
{
  void animalSound();
}
interface IAnimal2
{
  void animalSound2();
}

class Pig2 : IAnimal, IAnimal2
{
  public void animalSound()
  {
    Console.Write("The pig says: wee wee ");
  }
  public void animalSound2()
  {
    Console.WriteLine("wee wee weee");
  }
}
Pig2 pig2 = new();
pig2.animalSound();
pig2.animalSound2();
````
