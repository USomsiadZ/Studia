| `\"` | `"`         | Podwójny cudzysłów             |
| `\\` | `\`         | Backslash (ukośnik wsteczny)   |
| `\n` | nowa linia   | Przejście do nowej linii        |
| `\t` | tabulator    | Znak tabulacji                  |
| `\b` | backspace    | Cofnięcie o jeden znak (w tekście) |
