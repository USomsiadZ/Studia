# Metody na `string` w C#

| Nazwa | Typ | Opis | Przykład |
|------|-----|------|----------|
| `Length` | właściwość | Liczba znaków w stringu. | `int n = txt.Length;` |
| `Trim()` | metoda | Usuwa białe znaki z początku i końca. | `"  ala \n".Trim()` → `"ala"` |
| `Replace(old, new)` | metoda | Zamienia wszystkie wystąpienia `old` na `new`. | `"ala ma kota".Replace("a", "A")` |
| `Insert(index, value)` | metoda | Wstawia tekst w podanej pozycji. | `"abc".Insert(1, "X")` → `"aXbc"` |
| `Remove(start, count)` | metoda | Usuwa `count` znaków od `start`. | `"abcdef".Remove(2, 2)` → `"abef"` |
| `IndexOf(value)` | metoda | Indeks pierwszego wystąpienia (lub `-1`). | `"abrakadabra".IndexOf("ra")` → `2` |
| `Split(...)` | metoda | Dzieli string na tablicę wg separatorów. | `"a b c".Split(' ')` → `["a","b","c"]` |
| `string.Join(sep, kolekcja)` | **statyczna** | Łączy elementy kolekcji w jeden string. | `string.Join(", ", arr)` → `"a, b, c"` |
| `Substring(start, length)` | metoda | Zwraca fragment o danej długości. | `"abcdef".Substring(1, 3)` → `"bcd"` |
| `Contains(value)` | metoda | Sprawdza, czy tekst zawiera podciąg. | `"ala ma kota".Contains("ma")` → `true` |
| `StartsWith(value)` | metoda | Czy tekst zaczyna się od podciągu. | `"hello".StartsWith("he")` → `true` |
| `EndsWith(value)` | metoda | Czy tekst kończy się podciągiem. | `"hello".EndsWith("lo")` → `true` |
| `LastIndexOf(value)` | metoda | Indeks ostatniego wystąpienia. | `"abrakadabra".LastIndexOf("ra")` → `9` |
| `CompareTo(other)` | metoda | Porównanie leksykograficzne (do sortowania). | `"ala".CompareTo("ola")` → `< 0` |
| `string.Compare(a,b,…)` | **statyczna** | Zaawansowane porównanie dwóch stringów. | `string.Compare(a, b, true)` |
| `string.IsNullOrEmpty(s)` | **statyczna** | `true`, gdy `s` jest `null` lub `""`. | `string.IsNullOrEmpty(name)` |
| `string.IsNullOrWhiteSpace(s)` | **statyczna** | `true`, gdy `s` jest `null`, `""` lub samymi białymi znakami. | `string.IsNullOrWhiteSpace(input)` |
