```python
A = matrix([[0,6,3],[-1,-2,1]])
B = matrix([[6,0],[1,-2]])
X = A.solve_right(B)
show(X)
```

```python
a,b,c,d = var('a,b,c,d')
X = matrix(2,2,[a,b,c,d])
A = matrix([[1,0],[1,1]])
B = matrix([[1,0],[2,1]])
C = X*A - B*X
solution = solve(C.list(), a,b,c,d)
solution
```

Wyznaczyć rząd macierzy $A$ sprowadzając ją do postaci schodkowej:

$$
\begin{equation*}
A =
\left[\begin{array}{rrr}
2 & -1 & 1 & -1\ \
    1 & 1 & 1 & 1\ \
    2 & 0 & 2 & 2
\end{array}\right]
\end{equation*}
$$

```python
A = matrix([[2,-1,1,-1],[1,1,1,1],[2,0,2,2]])
rref_A = A.echelon_form()
rank_A = rref_A.rank()
show(rref_A)
show(rank_A)
```
