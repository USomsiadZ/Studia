### stdlib.h - Konwersja stringów na liczby

| Funkcja | Nagłówek | Zwraca | Opis |
|---------|----------|--------|------|
| `atoi` | `int atoi(const char *str);` | int | Konwertuje string na **int**. Pomija białe znaki na początku. Zwraca 0 jeśli błąd. |
| `strtol` | `long strtol(const char *str, char **endptr, int base);` | long | Konwertuje string na **long** z kontrolą błędów i określoną bazą. |
| `strtod` | `double strtod(const char *str, char **endptr);` | double | Konwertuje string na **double** z kontrolą błędów. |
* atol - long
* atof double 
**Przykłady:**
```c
char str1[] = "123";
int num = atoi(str1);              // num = 123

char str2[] = "  45.67";
double dbl = atof(str2);           // dbl = 45.67

char str3[] = "100 apples";
int result = atoi(str3);           // result = 100 (zatrzymuje się na spacji)

char str4[] = "abc";
int zero = atoi(str4);             // zero = 0 (błąd konwersji)
```


| Funkcja | Nagłówek | Zwraca | Opis |
|---------|----------|--------|------|
| `strcpy` | `char *strcpy(char *dest, const char *src);` | wskaźnik do dest | Kopiuje **Nie sprawdza długości bufora!** |
| `strncpy` | `char *strncpy(char *dest, const char *src, size_t n);` | wskaźnik do dest | Kopiuje n znaków |
| `strcat` | `char *strcat(char *dest, const char *src);` | wskaźnik do dest | dest+src+`'\0'` |
| `strcmp` | `int strcmp(const char *s1, const char *s2);` | wartość <0, 0 lub >0 | Porównuje dwa napisy leksykograficznie (alfabetycznie). |
| `strncmp` | `int strncmp(const char *s1, const char *s2, size_t n);` | wartość <0, 0 lub >0 | Porównuje co najwyżej n pierwszych znaków. |
| `strchr` | `char *strchr(const char *str, int c);` | wskaźnik lub NULL | Szuka ostatniego wystąpienia znaku c. |
| `strstr` | `char* strstr(const char* str, const char* substr);` | wskaźnik do pierwszego wystąpienia lub NULL | Szuka podnapisu w napisie (np. słowa w zdaniu). |
| `strtok` | `char *strtok(char *str, const char *delim);` | wskaźnik do tokena lub NULL | Dzieli napis na fragmenty (tokeny) według separatorów. |

| Funkcja | Nagłówek | Zwraca | Opis |
|---------|----------|--------|------|
| `memcpy` | `void *memcpy(void *dest, const void *src, size_t n);` | wskaźnik do dest | Kopiuje n bajtów z src do dest. **Nie obsługuje nakładających się obszarów pamięci!** |
| `memmove` | `void *memmove(void *dest, const void *src, size_t n);` | wskaźnik do dest | Kopiuje n bajtów z src do dest. **Bezpieczne dla nakładających się obszarów pamięci.** |
| `memset` | `void *memset(void *s, int c, size_t n);` | wskaźnik do s | Ustawia n pierwszych bajtów obszaru pamięci na wartość c. Często używane do wyzerowania: `memset(arr, 0, sizeof(arr));` |
| `memcmp` | `int memcmp(const void *s1, const void *s2, size_t n);` | wartość <0, 0 lub >0 | Porównuje n pierwszych bajtów dwóch obszarów pamięci. Zwraca 0 jeśli są identyczne. |
| `memchr` | `void *memchr(const void *s, int c, size_t n);` | wskaźnik do pierwszego wystąpienia lub NULL | Szuka pierwszego wystąpienia znaku c w pierwszych n bajtach obszaru pamięci. |


```c

int src[] = {1, 2, 3, 4, 5};
int desc[5];
memcpy(desc, src, sizeof(src));                    // Kopiowanie tablicy liczb całkowitych

char buffer[100] = "Hello";
memmove(buffer + 2, buffer, 10);                   // Bezpieczne kopiowanie z nakładającymi się obszarami

int arr[10];
memset(arr, 0, sizeof(arr));                       // Wyzerowanie tablicy

int data1[] = {1, 2, 3};
int data2[] = {1, 2, 3};
memcmp(data1, data2, sizeof(data1));              // Porównanie dwóch struktur

char buffer2[100] = "Test";
memchr(buffer2, 'A', 100);                          // Szukanie znaku 'A' w buforze
```

### Instrukcja continue
- Pomija pozostały kod w bieżącej iteracji

**Przykład:**
```c
#include <stdio.h>

int main(void) {
    for (int i = 1; i <= 5; i++) {
        if (i == 3) {
            continue;  // Pomija liczbę 3
        }
        printf("%d ", i);
    }
    return 0;
}
```
