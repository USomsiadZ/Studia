# Skonczylem na stronie - ORelly Jezyk c w pigułce - 64 

### Przenosi sie na poczatek lini z insertem
```c
#include <stdio.h>
printf("1");
printf("\r2");
```

### Korzystanie z jednej zmiennej w wielu plikach
```c
extern int a;// w data.h
int a = 1;// w count.c
printf("%d",a); // main.c - wypisze 1
```

### Widać enum w debugerze
```c
#include <stdio.h>
enum kolor {
    RED,
    GREEN,
    BLUE
};
enum kolor aktualny_kolor = RED;
// Enum -> int
void ustaw_aktualny(enum kolor k){
    aktualny_kolor = k;
}
int main() {
    ustaw_aktualny(2); // BLUE
    printf("%d", aktualny_kolor);
    return 0;
}
```

````c
#include <stdio.h>
// &a Jaki jest adres zmiennej w pamieci
// *a Co sie znajduje pod zmienna w pamieci
// printf %p do wypisywania adresu
int main() {
    int a = 1;      // normalna zmienna
    int *x = &a;
    
    printf("%p\n", &a); // wypisz adres
    printf("%d\n", *x); // wypisz adres

    return 0;
}
````
