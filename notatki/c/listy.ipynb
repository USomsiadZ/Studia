````c
#include <stdlib.h>
typedef struct Node {
    int value;
    struct Node* next;
} Node;

int main(){
    Node* head = NULL;
    Node* n = NULL;
    Node* p = NULL;
    
    n = calloc(1, sizeof(Node));
    n->value = 10;
    head = n;
    
    n = calloc(1, sizeof(Node));
    n->value = 20;
    head->next = n;
    p = n;
    
    n = calloc(1, sizeof(Node));
    n->value = 30;
    p->next = n;
    
    return 0;
}
````
