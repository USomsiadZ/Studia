### Tworzenie dynamicznych tablic 2d
```c
int** array = malloc(ROWS * sizeof(int*));
for (int i = 0; i < ROWS; i++) {
    array[i] = malloc(COLS * sizeof(int));
}
```
* array[i] to adres wiersza i.
* array[i][j] dostęp do j-tego elementu w wierszu o indeksie i.

```c 
// Równoważnie
*(*(array + i) + j)
```
#### Zwalnianie
```c
for (int i = 0; i < ROWS; i++) {
    free(array[i]); // zwolnienie i-tego wiersza
}
free(array); // zwolnienie tablicy wskaźników
```

#### **int zamiast return
```c
// **ptr, bo chcemy zmienić wskaźnik
void createArray(int **ptr, int size) {
    *ptr = (int*)malloc(size * sizeof(int));
}

int *arr = NULL;
createArray(&arr, 10);
```

### Gotowiec
```c
void create2D(int*** arr, int rows, int cols) {
    *arr = malloc(rows * sizeof(int*));
    for (int i = 0; i < rows; i++) {
        (*arr)[i] = malloc(cols * sizeof(int));
    }
}

void free2Dptr(int*** arr, int rows) {
    for (int i = 0; i < rows; i++) {
        free((*arr)[i]);
    }
    free(*arr);
    *arr = NULL;
}

//Wywołanie:
int** matrix = NULL;
create2D(&matrix, ROWS, COLS);
free2Dptr(&matrix, ROWS);


```
