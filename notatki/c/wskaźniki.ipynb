### <b>*p</b> - wartość zmiennej z adresu
### <b>&p</b> - adres zmiennej z wartosci

| Deklaracja | Mogę zmienić `*p` (dane)? | Mogę zmienić `p` (adres)? |
| :--- | :---: | :---: |
| `int *p` | ✅ TAK | ✅ TAK |
| `const int *p` | ❌ NIE | ✅ TAK |
| `int * const p` | ✅ TAK | ❌ NIE |
| `const int * const p` | ❌ NIE | ❌ NIE |

const int *p    - dane

int * const p   - wskaznik

Wskaźnik adresu dowolnego typu
```c
void *p;
```

````c
// alokowanie pamięci
int *p = malloc(10 * sizeof(*p));
int *p = calloc(10, sizeof(*p));
int *p = realloc(p, 20 * sizeof(*p));
free(p);
p = NULL;
````

````c
// używanie p->x
#include <stdio.h>
#include <stdlib.h>
struct Galaktyka{
    int x;
    int y;
    int z;
};
int main(){
    struct Galaktyka star = {10,12,3000};
    struct Galaktyka *p = &star;
    printf("%d\n",(*p).x);
    printf("%d\n",p->x);
    scanf("%d",&p->x);
    printf("%d\n",p->x);
    return 0;
}
````

````c
// Składając struktury
#include <stdio.h>
#include <stdlib.h>
typedef struct{
    int x;
    int y;
} Point;

typedef struct{
    Point planeta[10];
} Planeta;

int main(){
    Planeta uklad[10];
    uklad[1].planeta[1].x = 1;
    return 0;
}
````
