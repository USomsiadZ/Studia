| Zmienna | Opis funkcji |
|---------|---------------|
| $PATH | Najważniejsza zmienna. Zawiera listę katalogów oddzielonych dwukropkiem :, w których system szuka plików wykonywalnych. Jeśli wpiszesz ls, system sprawdza po kolei katalogi z $PATH, aby znaleźć ten program. |
| $HOME | Ścieżka do katalogu domowego aktualnego użytkownika (np. /home/imrozek). |
| $USER / $USERNAME | Nazwa aktualnie zalogowanego użytkownika. |
| $PWD | Przechowuje ścieżkę do aktualnego katalogu roboczego (wynik polecenia pwd). |
| $SHELL | Ścieżka do aktualnie używanej powłoki (np. /bin/bash). |
| $LANG | Definiuje język i kodowanie znaków w systemie (np. pl_PL.UTF-8). |
| $TERM | Typ terminala używanego do wyświetlania znaków (np. xterm). |
