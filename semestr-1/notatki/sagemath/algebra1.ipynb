## 1. Równania macierzowe z niewiadomą X

### Przykład 1: Równanie typu (A + X)/2 = B

````python
A = matrix([[-2, 2, -1], [4, 0, 1]])
B = matrix([[0, 1, -3], [2, 1, 0]])
X = (2*B - A)
print("Rozwiązanie X:")
print(X)
````

### Przykład 2: Równanie A*X = B

````python
A = matrix([[0, 6, 3], [-1, -2, 1]])
B = matrix([[6, 0], [1, -2]])
X = A.solve_right(B)

print("Macierz A:")
print(A)
print("\nMacierz B:")
print(B)
print("\nRozwiązanie X:")
print(X)
````

### Przykład 3: Równanie X*A - B*X = 0

````python
a, b, c, d = var('a b c d')
X = Matrix([[a, b], [c, d]])
A = Matrix([[1, 0], [1, 1]])
B = Matrix([[1, 0], [2, 1]])

# Rozwiązanie: X*A - B*X = 0
rozw = solve([c == 0 for c in (X*A - B*X).list()], [a, b, c, d])
print("Rozwiązanie:")
rozw
````

## 2. Postać schodkowa i rząd macierzy

````python
A = matrix([[2, -1, 1, -1], [1, 1, 1, 1], [2, 0, 2, 2]])

# Postać schodkowa
postac_schodkowa = A.echelon_form()
rzad = A.rank()

print(f"Rząd macierzy: {rzad}")
print("\nPostać schodkowa:")
postac_schodkowa
````

## 3. Macierz odwrotna

### Metoda 1: Z definicji (rozwiązanie równania A*X = I)

````python
p, q, r, s, t, w, u, x, z = var('p q r s t w u x z')
A = matrix([[1, 0, 2], [2, 1, 1], [2, 1, 0]])
X = matrix([[p, q, r], [s, t, w], [u, x, z]])
I = identity_matrix(3)

# Rozwiązanie A*X = I
roz = solve([c == 0 for c in (A*X - I).list()], [p, q, r, s, t, w, u, x, z])
roz2 = X.subs(roz)

print("Macierz odwrotna:")
roz2
````

### Metoda 2: Operacje elementarne

````python
A = matrix([[1, 0, 2], [2, 1, 1], [2, 1, 0]])
I = identity_matrix(3)

# Metoda macierzy rozszerzonej [A|I]
rozszerzona = A.augment(I, subdivide=True).echelon_form()
print("Macierz rozszerzona [A|I] w postaci schodkowej:")
print(rozszerzona)

# Lub bezpośrednio:
X = A.solve_right(I)
print("\nMacierz odwrotna:")
X
````

## 4. Postać bazowa (zredukowana)

````python
A = matrix([[1, -1, 2], [-2, 3, 1], [5, -7, -4]])

postac_bazowa = A.rref()
rzad = postac_bazowa.rank()

print(f"Rząd macierzy: {rzad}")
print("\nPostać bazowa (RREF):")
postac_bazowa
````

## 5. Rozwiązywanie układów równań

### Metoda 1: Używając RREF i kolumn jednostkowych

````python
A = matrix([[2, -1, 1], [-4, -12, 1], [3, 3, 1]])
B = vector([1, 2, 3])

# Macierz rozszerzona [A|B]
U = A.augment(B, subdivide=True).rref()
print("Postać bazowa [A|B]:")
print(U)
````

### Metoda 2: Używając solve

````python
A = matrix([[2, -1, 1], [-4, -12, 1], [3, 3, 1]])
B = vector([1, 2, 3])
x, y, z = var('x y z')
X = vector([x, y, z])

rozwiazanie = solve([c == 0 for c in (A*X - B).list()], [x, y, z])
print("Rozwiązanie:")
rozwiazanie
````

## 6. Permutacje

### Rząd permutacji (order)

````python
s = PermutationGroupElement([2, 3, 1, 5, 4, 8, 10, 9, 7, 6])

# Rząd permutacji - najmniejsza liczba, po której s^n = e
rzad = s.order()
print(f"Permutacja: {s}")
print(f"Rząd permutacji: {rzad}")

# Alternatywnie - LCM rzędów cykli
rzad_lcm = lcm([c.order() for c in s.cycles()])
print(f"Rząd (przez LCM cykli): {rzad_lcm}")
````

### Operacje na permutacjach

````python
s = PermutationGroupElement([2, 5, 3, 1, 4])
t = PermutationGroupElement([4, 5, 2, 1, 3])

print("Permutacja s:", Permutation(s))
print("Permutacja t:", Permutation(t))
print("\nIloczyn t*s:", Permutation(t*s))
print("Iloczyn s*t:", Permutation(s*t))
print("\nPotęga s^4:", Permutation(s^4))
print("Odwrotność t^(-1):", Permutation(t^(-1)))
print("\nDuża potęga t^2021:", Permutation(t^2021))
````

### Parzystość permutacji

````python
a = PermutationGroupElement([2, 3, 1, 5, 4, 8, 10, 9, 7, 6])

print(f"Permutacja: {a}")
print(f"Znak (parzystość): {a.sign()}")
print("\nZnak = 1 oznacza permutację parzystą")
print("Znak = -1 oznacza permutację nieparzystą")
````

### Sprzężenie permutacji (zadanie Rajmunda)

````python
# Wzór: t = s^(-1) * a * s^n
a = PermutationGroupElement([9, 1, 4, 2, 6, 5, 3, 8, 7])
s = PermutationGroupElement([4, 7, 6, 5, 1, 9, 2, 3, 8])

t = Permutation((s^(-1)) * a * (s^26))
print("Wynik:")
t
````
