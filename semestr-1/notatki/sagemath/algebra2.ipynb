## 1. Operacje na macierzach

### Podstawowe funkcje SageMath
```python
A = matrix([[...]...])      # tworzenie macierzy
A + B                       # dodawanie
A * B                       # mnożenie
A.transpose() lub A.T       # transpozycja
A.echelon_form()            # postać schodkowa
A.rref()                    # postać bazowa
A.rank()                    # rząd macierzy
A.inverse() lub A^(-1)      # macierz odwrotna
identity_matrix(n)          # macierz jednostkowa nxn
```

### Przykład: Dodawanie i mnożenie macierzy

````sage
# Dodawanie macierzy
A = matrix([[2, -1, 3], [1, 0, 2]])
B = matrix([[1, -2, 0], [-3, 4, -1]])
wynik = A + 2*B
show(wynik)
````

````sage
# Mnożenie macierzy
A = matrix([[3, 0, 2, 0], [0, 1, 2, 1], [2, 3, 0, 0]])
B = matrix([[1, -2, 2], [2, -1, 1], [-1, 1, -2], [2, 2, -1]])
C = matrix([[-2, 0, -3], [0, 6, -3], [5, -2, 8]])

wynik = A*B + 3*C
show(wynik)
````

### Przykład: Transpozycja

````sage
A = matrix([[1, 3], [2, -3], [1, 2]])
A_T = A.transpose()
# lub krócej: A_T = A.T
show("Macierz A:", A)
show("Macierz A^T:", A_T)
````

## 2. Równania macierzowe

### Typ 1: Równania liniowe (2X - A = B)

````sage
# 2X - A = B  =>  X = (B + A) / 2
A = matrix([[-2, 2, -1], [4, 0, 1]])
B = matrix([[0, 1, -3], [2, 1, 0]])
X = (B + A) / 2

# Weryfikacja
print("Sprawdzenie:", 2*X - A == B)
show(X)
````

### Typ 2: Równania A*X = B

````sage
# Metoda solve_right
A = matrix([[0, 6, 3], [-1, -2, 1]])
B = matrix([[6, 0], [1, -2]])
X = A.solve_right(B)

print("Sprawdzenie:", A*X == B)
show(X)
````

### Typ 3: Równania z niewiadomą macierzą X*A - B*X = 0

````sage
# Używamy zmiennych symbolicznych
a, b, c, d = var('a b c d')
X = matrix([[a, b], [c, d]])
A = matrix([[1, 0], [1, 1]])
B = matrix([[1, 0], [2, 1]])

# Rozwiązanie równania X*A - B*X = 0
C = X*A - B*X
rozwiazanie = solve(C.list(), a, b, c, d)
show(rozwiazanie)
````

## 3. Rząd macierzy i postać schodkowa

### Metoda echelon_form()

````sage
A = matrix([[2, -1, 1, -1], [1, 1, 1, 1], [2, 0, 2, 2]])

# Postać schodkowa
postac_schodkowa = A.echelon_form()
rzad = A.rank()

print(f"Rząd macierzy: {rzad}")
show("Postać schodkowa:", postac_schodkowa)
````

### Przykład z większą macierzą

````sage
A = matrix([
    [5, 28, 1, 4, -5, 1, 0],
    [0, 0, -3, 0, -1, 1, -2],
    [41, 1, 1, 0, -1, 0, 0],
    [1, 0, 4, -1, 2, -11, 1],
    [0, 0, -1, 4, -6, -2, 0],
    [1, -1, 13, 1, -2, 1, 1],
    [-3, -5, 0, -1, 1, -1, 2]
])

rref_a = A.echelon_form()
rzad = A.rank()

print(f"Rząd: {rzad}")
show(rref_a)
````

## 4. Postać bazowa (RREF)

### Metoda rref() - Zredukowana postać schodkowa

````sage
A = matrix([[1, -1, 2], [-2, 3, 1], [5, -7, -4]])

# Postać bazowa (RREF)
postac_bazowa = A.rref()
rzad = postac_bazowa.rank()

print(f"Rząd: {rzad}")
show("RREF:", postac_bazowa)
````

## 5. Macierz odwrotna

### Metoda 1: Z definicji (A*X = I)

````sage
# Dla macierzy 2x2
p, q, r, t = var('p q r t')
A = matrix([[3, 1], [5, 2]])
X = matrix([[p, q], [r, t]])
I = identity_matrix(2)

# Rozwiązanie A*X = I
rozwiazanie = solve([c == 0 for c in (A*X - I).list()], [p, q, r, t])
X_inv = matrix([[s.rhs() for s in rozwiazanie[0][:2]], 
                [s.rhs() for s in rozwiazanie[0][2:]]])

print("Sprawdzenie:", A*X_inv == I)
show(X_inv)
````

### Metoda 2: Używając metody inverse()

````sage
A = matrix([[3, 1], [5, 2]])
A_inv = A.inverse()
# lub: A_inv = A^(-1)

print("Sprawdzenie:", A*A_inv == identity_matrix(2))
show(A_inv)
````

### Metoda 3: Operacje elementarne [A|I] → [I|A^(-1)]

````sage
A = matrix([[1, 0, 2], [2, 1, 1], [2, 1, 0]])
I = identity_matrix(3)

# Macierz rozszerzona [A|I]
rozszerzona = A.augment(I, subdivide=True).echelon_form()
show("[A|I] w postaci schodkowej:", rozszerzona)

# Lub bezpośrednio:
A_inv = A.inverse()
show("A^(-1):", A_inv)
````

## 6. Układy równań liniowych

### Metoda 1: Eliminacja Gaussa (macierz rozszerzona)

````sage
# Układ równań: 3x + 2y + 4z + t = 1
#               3x + y + 6z + 4t = 2
#               2x + 2y + 2z + 2t = 0

A = matrix([[3, 2, 4, 1], [3, 1, 6, 4], [2, 2, 2, 2]])
b = vector([1, 2, 0])

# Macierz rozszerzona [A|b]
rozszerzona = A.augment(b, subdivide=True)
rref = rozszerzona.rref()
show("RREF [A|b]:", rref)
````

### Metoda 2: Funkcja solve()

````sage
# Definicja zmiennych
var('x y z t')

# Układ równań
eq1 = 3*x + 2*y + 4*z + t == 1
eq2 = 3*x + y + 6*z + 4*t == 2
eq3 = 2*x + 2*y + 2*z + 2*t == 0

# Rozwiązanie
rozwiazanie = solve([eq1, eq2, eq3], x, y, z, t)
show(rozwiazanie)
````

### Metoda 3: Kolumny jednostkowe (RREF)

````sage
# Układ: 2x - y + z = 1
#        -4x - 12y + z = 2
#        3x + 3y + z = 3

A = matrix([[2, -1, 1], [-4, -12, 1], [3, 3, 1]])
B = vector([1, 2, 3])

# RREF macierzy rozszerzonej
U = A.augment(B, subdivide=True).rref()
show("Rozwiązanie w postaci RREF:", U)
````

## 7. Permutacje

### Tworzenie permutacji

````sage
# Permutacja σ = (1 2 3 4 5) → (2 3 1 5 4)
sigma = PermutationGroupElement([2, 3, 1, 5, 4])
print("Permutacja:", sigma)
print("Cykle:", sigma.cycles())
````

### Rząd permutacji

````sage
s = PermutationGroupElement([2, 3, 1, 5, 4, 8, 10, 9, 7, 6])

# Rząd permutacji - najmniejsza liczba k, dla której s^k = e
rzad = s.order()
print(f"Permutacja: {s}")
print(f"Rząd: {rzad}")

# Alternatywnie: LCM rzędów cykli
rzad_lcm = lcm([c.order() for c in s.cycles()])
print(f"Rząd (LCM cykli): {rzad_lcm}")
````

### Operacje na permutacjach

````sage
s = PermutationGroupElement([2, 5, 3, 1, 4])
t = PermutationGroupElement([4, 5, 2, 1, 3])

print("s:", Permutation(s))
print("t:", Permutation(t))
print("\nIloczyn t*s:", Permutation(t*s))
print("Iloczyn s*t:", Permutation(s*t))
print("\nPotęga s^4:", Permutation(s^4))
print("Odwrotność t^(-1):", Permutation(t^(-1)))
print("\nDuża potęga t^2021:", Permutation(t^2021))
````

### Parzystość permutacji

````sage
a = PermutationGroupElement([2, 3, 1, 5, 4, 8, 10, 9, 7, 6])

print(f"Permutacja: {a}")
print(f"Znak: {a.sign()}")
print("\nZnak = 1  → permutacja parzysta")
print("Znak = -1 → permutacja nieparzysta")
````

### Sprzężenie permutacji

````sage
# Wzór: τ = σ^(-1) * a * σ^n
a = PermutationGroupElement([9, 1, 4, 2, 6, 5, 3, 8, 7])
s = PermutationGroupElement([4, 7, 6, 5, 1, 9, 2, 3, 8])

tau = Permutation((s^(-1)) * a * (s^26))
print("τ =", tau)
````

## 9. Liczby zespolone

### Podstawowe operacje

````sage
# W SageMath: I oznacza jednostkę urojoną
z = 3 - 4*I
w = 5 + 12*I

print(f"z = {z}")
print(f"w = {w}")
print(f"\nRe(z) = {z.real()}")
print(f"Im(z) = {z.imag()}")
print(f"|z| = {abs(z)}")
print(f"z̄ (sprzężenie) = {z.conjugate()}")
````

### Rozwiązywanie równań zespolonych

````sage
# Równanie: z^2 + z + 1 = 0
z = var('z')
eq = z**2 + z + 1 == 0
rozwiazania = solve(eq, z)
show("Rozwiązania:", rozwiazania)
````

### Równania z |z| (moduł)

````sage
# Równanie: |z| + z = 8 + 4i
x, y = var('x y')
z = x + y*I

eq = abs(z) + z == 8 + 4*I

# Rozdzielamy część rzeczywistą i urojoną
solution = solve([
    (sqrt(x**2 + y**2) + x) == 8,  # część rzeczywista
    y == 4                          # część urojona
], [x, y])
show(solution)
````

### Wielokąty foremne na płaszczyźnie zespolonej

````sage
# Siedmiokąt foremny o wierzchołku w (1,1)
# Środek w (0,0)
u = 1 + 1*I  # punkt (1,1)
n = 7        # liczba boków

# Pierwiastki n-tego stopnia z liczby u^n
wierzcholki = [u * e^(2*pi*I*k/n) for k in range(n)]
rys = polygon(wierzcholki, fill=False)
rys
````

### Obrót figury o kąt α

````sage
# Obrót trójkąta o π/10 zgodnie z ruchem wskazówek zegara
u = -2 + 5*I  # wierzchołek
alpha = -pi/10  # ujemny kąt = zgodnie z ruchem wskazówek

# Wierzchołki trójkąta
trojkat = [u * e^(2*pi*I*k/3) for k in range(3)]

# Obrót: mnożenie przez e^(i*alpha)
trojkat_obrocony = [w * e^(I*alpha) for w in trojkat]

rys = polygon(trojkat, fill=False, color='blue') + \
      polygon(trojkat_obrocony, fill=False, color='red')
rys
````

## 10. Liniowa niezależność wektorów

### Metoda 1: Wyznacznik (dla macierzy kwadratowych)

````sage
# Wektory w R^3: (1,0,3), (0,2,0), (1,-1,3)
A = matrix(QQ, [[1, 0, 3], [0, 2, 0], [1, -1, 3]])

det_A = A.det()
print(f"Wyznacznik: {det_A}")

if det_A != 0:
    print("Wektory są liniowo niezależne")
else:
    print("Wektory są liniowo zależne")
````

### Metoda 2: Rząd macierzy

````sage
# Wektory w R^5: (1,2,3,-1,0), (5,2,0,1,1), (6,4,3,0,1)
A = matrix(QQ, [[1, 2, 3, -1, 0], [5, 2, 0, 1, 1], [6, 4, 3, 0, 1]])

rzad = A.rank()
liczba_wektorow = A.nrows()

print(f"Rząd macierzy: {rzad}")
print(f"Liczba wektorów: {liczba_wektorow}")

if rzad == liczba_wektorow:
    print("Wektory są liniowo niezależne")
else:
    print("Wektory są liniowo zależne")
````

### Metoda 3: Pivot rows

````sage
A = matrix(QQ, [[1, 2, 3], [5, 2, 0], [1, -1, 3]])

# Indeksy wierszy wiodących
pivot_rows = A.pivot_rows()
print(f"Wiersze wiodące: {pivot_rows}")
print(f"Liczba wierszy wiodących: {len(pivot_rows)}")

if len(pivot_rows) == A.nrows():
    print("Wektory są liniowo niezależne")
else:
    print("Wektory są liniowo zależne")
````

### Metoda 4: Solve (kombinacja liniowa = 0)

````sage
# Sprawdzamy, czy istnieje nietrywialna kombinacja liniowa = 0
x, y, z = var('x y z', domain='real')
v1, v2, v3 = vector([1, 2, 3]), vector([5, 2, 0]), vector([1, -1, 3])

rozwiazanie = solve([s == 0 for s in x*v1 + y*v2 + z*v3], [x, y, z])
print("Rozwiązanie równania x*v1 + y*v2 + z*v3 = 0:")
show(rozwiazanie)

# Jeśli jedyne rozwiązanie to x=y=z=0, wektory są niezależne
````

### Liniowa niezależność wielomianów

````sage
# Wielomiany: x^2+1, 3x-2, 3x^2+7 w R[x]_2
# Reprezentujemy jako wektory współczynników [a0, a1, a2] dla a0 + a1*x + a2*x^2

# x^2+1    → [1, 0, 1]
# 3x-2     → [-2, 3, 0]
# 3x^2+7   → [7, 0, 3]

A = matrix(QQ, [[1, 0, 1], [-2, 3, 0], [7, 0, 3]])
rzad = A.rank()

print(f"Rząd: {rzad}")
if rzad == 3:
    print("Wielomiany są liniowo niezależne")
else:
    print("Wielomiany są liniowo zależne")
````

## Podsumowanie wzorów i technik

### Macierze
- **Rząd**: `A.rank()`
- **Postać schodkowa**: `A.echelon_form()`
- **RREF**: `A.rref()`
- **Odwrotność**: `A.inverse()` lub `A^(-1)`
- **Transpozycja**: `A.transpose()` lub `A.T`
- **Wyznacznik**: `A.det()`

### Równania
- **A*X = B**: `X = A.solve_right(B)`
- **Układy równań**: `solve([eq1, eq2, ...], [x, y, ...])`
- **Macierz rozszerzona**: `A.augment(b, subdivide=True).rref()`

### Permutacje
- **Tworzenie**: `PermutationGroupElement([...])`
- **Rząd**: `s.order()`
- **Parzystość**: `s.sign()`
- **Cykle**: `s.cycles()`

### Liczby zespolone
- **Jednostka urojona**: `I`
- **Moduł**: `abs(z)`
- **Sprzężenie**: `z.conjugate()`
- **Postać wykładnicza**: `r * e^(I*theta)`
- **Obrót**: mnożenie przez `e^(I*alpha)`

### Liniowa niezależność
- **Wyznacznik**: dla macierzy kwadratowych
- **Rząd**: `A.rank()` vs liczba wektorów
- **Pivot rows**: `A.pivot_rows()`
- **Solve**: sprawdzenie, czy kombinacja liniowa = 0
