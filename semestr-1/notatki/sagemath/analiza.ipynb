## 4. Asymptoty

### Asymptota pozioma

Dla funkcji $f(x)$, asymptota pozioma to prosta $y = L$, gdzie:
$$\lim_{x \to \pm\infty} f(x) = L$$

```python
f(x) = 1/x

# Obliczenie granicy
L = lim(f(x), x=oo)

# Narysowanie asymptoty poziomej
plot(L, (x, -10, 10), color='red', linestyle='--')
```

### Asymptota pionowa

Dla funkcji $f(x)$, asymptota pionowa w punkcie $x_0$ występuje gdy:
$$\lim_{x \to x_0^-} f(x) = \pm\infty \quad \text{lub} \quad \lim_{x \to x_0^+} f(x) = \pm\infty$$

```python
f(x) = 1/x

# Sprawdzenie granicy jednostronnej
g1 = lim(f(x), x=0, dir='-')
g2 = lim(f(x), x=0, dir='+')

# Rysowanie asymptoty pionowej
line([(0, -10), (0, 10)], color='red', linestyle='--')
```

### Asymptota ukośna

Dla funkcji $f(x)$, asymptota ukośna $y = ax + b$ jest wyznaczana wzorami:
$$a = \lim_{x \to \infty} \frac{f(x)}{x}$$
$$b = \lim_{x \to \infty} (f(x) - ax)$$

```python
f(x) = x^2/x  # przykład

# Obliczenie współczynników
a = lim(f(x)/x, x=oo)
b = lim(f(x) - a*x, x=oo)

# Równanie asymptoty
asymptota = a*x + b

# Rysowanie
plot(asymptota, (x, -10, 10), color='red', linestyle='--')
```

## 5. Szeregi

### Definicja szeregu

Szereg liczbowy to nieskończona suma liczb:
$$S = \sum_{n=1}^{\infty} a_n = a_1 + a_2 + a_3 + \cdots$$

### Ciąg sum częściowych

$$s_n = \sum_{k=1}^n a_k = a_1 + \cdots + a_n$$

Suma szeregu to granica ciągu sum częściowych:
$$\lim_{n \to \infty} s_n = S$$

```python
var('k')
a(n) = 1/(n*(n+1))

# Ciąg sum częściowych
s(n) = sum(a(k), k, 1, n)

# Suma szeregu
S = lim(s(n), n=oo)
print(f'Suma szeregu: {S}')
```

### Warunek konieczny zbieżności

Jeżeli szereg $\sum_{n=1}^\infty a_n$ jest zbieżny, to:
$$\lim_{n \to \infty} a_n = 0$$

```python
a(n) = 1/n^2

g = lim(a(n), n=oo)

if g == 0:
    print('Warunek konieczny spełniony - szereg może być zbieżny')
else:
    print('Warunek konieczny niespełniony - szereg rozbieżny')
```

### Szereg potęgowy

Szereg potęgowy ma postać:
$$\sum_{n=0}^\infty a_n (x-x_0)^n$$

Promień zbieżności:
$$R = \lim_{n \to \infty} \left| \frac{a_n}{a_{n+1}} \right|$$

Przedział zbieżności: $(x_0 - R,\ x_0 + R)$

```python
a(n) = (-1)^n / ((n+1)^2 * 2^n)
x0 = 3

# Promień zbieżności
R = lim(abs(a(n)/a(n+1)), n=oo)

print(f'Przedział zbieżności: ({x0 - R}, {x0 + R})')
```

## 6. Szeregi Fouriera i Wielomiany Taylora

### Wielomiany Taylora

Wielomian Taylora stopnia $n$ funkcji $f(x)$ o środku w punkcie $x_0$:

```python
f(x) = sin(x)

# Wielomian Taylora stopnia 3 o środku w 0
W3 = taylor(f(x), x, 0, 3)

# Wielomian Taylora stopnia 5 o środku w π
W5 = taylor(f(x), x, pi, 5)

print(f'W3 = {W3}')
print(f'W5 = {W5}')
```

### Wizualizacja wielomianów Taylora

```python
# Porównanie funkcji i jej aproksymacji
rys = plot(f, (x, -4, 7), color='red', thickness=3, title='Wielomiany Taylora')
rys += plot(W3, (x, -4, 7), color='blue', legend_label='Stopień 3')
rys += plot(W5, (x, -4, 7), color='green', legend_label='Stopień 5')
rys
```

### Zadanie 2: Badanie ciągłości

Zbadaj ciągłość funkcji w punkcie $x_0$:

```python
f(x) = (x^3 + 3*x^2 + 2*x) / abs(x+2)

# Granice jednostronne
g_left = lim(f(x), x=-2, dir='-')
g_right = lim(f(x), x=-2, dir='+')

if g_left == g_right:
    print('Funkcja jest ciągła w x=-2')
else:
    print('Funkcja ma nieciągłość w x=-2')
```
