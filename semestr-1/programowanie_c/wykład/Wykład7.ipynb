```c
#include <stdio.h>
#include <stdlib.h>

int main() {
    // Przykład użycia malloc - alokacja pamięci dla tablicy 5 liczb całkowitych
    int *tablica;
    int rozmiar = 5;
    
    // Alokacja pamięci
    tablica = malloc(rozmiar * sizeof(int));
    
    // Sprawdzenie czy alokacja się powiodła
    if (tablica == NULL) {
        printf("Błąd alokacji pamięci!\n");
        return 1;
    }
    
    // Wypełnienie tablicy wartościami
    for (int i = 0; i < rozmiar; i++) {
        tablica[i] = i * 10;
    }
    
    // Wyświetlenie wartości
    printf("Zawartość tablicy:\n");
    for (int i = 0; i < rozmiar; i++) {
        printf("tablica[%d] = %d\n", i, tablica[i]);
    }
    
    // Zwolnienie pamięci
    free(tablica);
    
    return 0;
}
```
